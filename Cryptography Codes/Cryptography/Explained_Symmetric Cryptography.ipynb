# 1. **Caesar Cipher Algorithm (Simple Steps)**  

## **Concept**  
The **Caesar Cipher** shifts each letter forward (for encryption) or backward (for decryption) by a fixed number of places.  

## **Algorithm Steps**  
1. **Take user input** â†’ Get the text to encrypt and the shift value.  
2. **Loop through each character**:  
   - If itâ€™s a **letter** (A-Z or a-z), shift it forward by the given amount.  
   - If itâ€™s **not a letter** (like spaces, numbers, or punctuation), leave it unchanged.  
3. **Encryption formula**:  
   - Convert the letter to its ASCII number.  
   - Shift it forward by the given amount.  
   - Convert it back to a letter.  
4. **For decryption**, shift **backward** instead.  

---

## **C++ Implementation**  

```cpp
#include <iostream>
using namespace std;

string caesarCipher(string text, int shift) {
    string result = "";
    for (char c : text) {
        if (isalpha(c)) {
            char base = isupper(c) ? 'A' : 'a';
            result += (char)((c - base + shift) % 26 + base);
        } else {
            result += c; // Keep non-letters unchanged
        }
    }
    return result;
}

int main() {
    string text;
    int shift;
    
    cout << "Enter text: ";
    getline(cin, text);
    
    cout << "Enter shift key: ";
    cin >> shift;
    
    string encrypted = caesarCipher(text, shift);
    cout << "Encrypted: " << encrypted << endl;
    
    string decrypted = caesarCipher(encrypted, -shift);
    cout << "Decrypted: " << decrypted << endl;

    return 0;
}


# 2.**Monoalphabetic Substitution Cipher (Simple Steps)**  

## **Concept**  
- Each letter in the alphabet is replaced with a **fixed, random letter**.  
- Only the sender and receiver know the **key mapping**.  
- This method is vulnerable to **frequency analysis** but is stronger than the Caesar cipher.  

## **Algorithm Steps**  
1. **Define a key** â†’ A dictionary that maps each letter to a unique random letter.  
2. **Reverse the key** â†’ This allows decryption by mapping the encrypted letters back to their originals.  
3. **Encryption Process**:  
   - Convert the text to lowercase.  
   - Replace each letter using the key dictionary.  
   - Keep non-alphabet characters unchanged.  
4. **Decryption Process**:  
   - Use the reversed key to get back the original letters.  

---

## **C++ Implementation**  

```cpp
#include <iostream>
#include <unordered_map>
using namespace std;

// Define a fixed key (letter mappings for encryption)
unordered_map<char, char> fixed_key = {
    {'a', 'r'}, {'b', 'y'}, {'c', 'd'}, {'d', 'q'}, {'e', 'w'}, {'f', 'u'}, {'g', 'k'},
    {'h', 'j'}, {'i', 'v'}, {'j', 'n'}, {'k', 'x'}, {'l', 'b'}, {'m', 'p'}, {'n', 'o'},
    {'o', 's'}, {'p', 'c'}, {'q', 'e'}, {'r', 'z'}, {'s', 'f'}, {'t', 'l'}, {'u', 'g'},
    {'v', 'a'}, {'w', 'm'}, {'x', 't'}, {'y', 'h'}, {'z', 'i'}
};

// Create a reverse key for decryption
unordered_map<char, char> reverse_key;
void generateReverseKey() {
    for (auto pair : fixed_key) {
        reverse_key[pair.second] = pair.first;
    }
}

// Function to encrypt text
string encrypt(string text, unordered_map<char, char> key) {
    string result = "";
    for (char c : text) {
        if (isalpha(c)) {
            result += key[tolower(c)];
        } else {
            result += c; // Keep non-alphabet characters unchanged
        }
    }
    return result;
}

// Function to decrypt text
string decrypt(string text, unordered_map<char, char> key) {
    return encrypt(text, key); // Same function, but with the reverse key
}

int main() {
    generateReverseKey();

    string text;
    cout << "Enter text to encrypt: ";
    getline(cin, text);

    string encrypted_text = encrypt(text, fixed_key);
    cout << "Encrypted: " << encrypted_text << endl;

    string decrypted_text = decrypt(encrypted_text, reverse_key);
    cout << "Decrypted: " << decrypted_text << endl;

    return 0;
}


# 3. **XOR Cipher (Simple Steps)**  

## **Concept**  
- The **XOR cipher** encrypts text by applying the **XOR (âŠ•) operation** between each character and a fixed key.  
- XOR is **reversible**, meaning encryption and decryption use the **same function**.  
- Commonly used in **stream ciphers** and **low-level encryption** techniques.  

## **Algorithm Steps**  
1. **Choose a secret key** (an integer value that must be shared securely).  
2. **Encryption Process**:  
   - Convert each character to its ASCII value.  
   - Apply the **XOR operation** with the key.  
   - Convert the result back to a character.  
3. **Decryption Process**:  
   - Use the same XOR operation on the encrypted text (XOR is self-reversing).  
   - Convert back to readable text.  

---

## **C++ Implementation**  

```cpp
#include <iostream>
using namespace std;

// XOR Cipher function (Same for Encryption & Decryption)
string xorCipher(string text, char key) {
    string result = "";
    for (char c : text) {
        result += c ^ key;  // XOR operation
    }
    return result;
}

int main() {
    string text;
    char key = 42;  // Secret Key (should be securely shared)

    cout << "Enter text to encrypt: ";
    getline(cin, text);

    // Encrypt
    string encrypted_text = xorCipher(text, key);
    cout << "Encrypted (as bytes): ";
    for (char c : encrypted_text) {
        cout << "\\x" << hex << (int)(unsigned char)c;  // Show as hex values
    }
    cout << endl;

    // Decrypt
    string decrypted_text = xorCipher(encrypted_text, key);
    cout << "Decrypted: " << decrypted_text << endl;

    return 0;
}


# 4.**VigenÃ¨re Cipher (Simple Steps)**  

## **Concept**  
- The **VigenÃ¨re Cipher** is an extension of the Caesar Cipher but uses a **keyword** instead of a fixed shift.  
- Each letter is shifted by a different amount, determined by the corresponding letter in the **keyword**.  
- It is **resistant to simple frequency analysis** but can be broken using the **Kasiski examination**.  

## **Algorithm Steps**  
1. **Take user input** â†’ Get the text and a **keyword**.  
2. **Repeat the keyword** so that it matches the length of the text.  
3. **Loop through each character in the text**:  
   - If itâ€™s a **letter**, calculate the shift from the keyword.  
   - Apply a **forward shift** for encryption.  
   - Apply a **backward shift** for decryption.  
   - Keep non-alphabet characters unchanged.  
4. **Return the final encrypted/decrypted text**.  

---

## **C++ Implementation**  

```cpp
#include <iostream>
using namespace std;

// VigenÃ¨re Cipher function (Encryption & Decryption)
string vigenereCipher(string text, string key, bool encrypt = true) {
    string result = "";
    int key_length = key.length();

    for (size_t i = 0, j = 0; i < text.length(); i++) {
        char c = text[i];
        if (isalpha(c)) {
            char base = isupper(c) ? 'A' : 'a';
            int shift = tolower(key[j % key_length]) - 'a'; // Get shift from keyword
            if (!encrypt) shift = -shift; // Reverse shift for decryption

            result += char((c - base + shift + 26) % 26 + base); // Apply shift
            j++; // Move to next key letter
        } else {
            result += c; // Keep non-alphabet characters unchanged
        }
    }
    return result;
}

int main() {
    string text, key;

    cout << "Enter text: ";
    getline(cin, text);

    cout << "Enter keyword: ";
    cin >> key;

    // Encrypt
    string encrypted_text = vigenereCipher(text, key);
    cout << "Encrypted: " << encrypted_text << endl;

    // Decrypt
    string decrypted_text = vigenereCipher(encrypted_text, key, false);
    cout << "Decrypted: " << decrypted_text << endl;

    return 0;
}


# 5.**Playfair Cipher (Simple Steps)**  

## **Concept**  
- The **Playfair Cipher** encrypts text using **digraphs (pairs of letters)**.  
- A **5Ã—5 matrix** (generated from a secret key) determines letter mappings.  
- The letter **J is replaced with I** to fit in the matrix.  
- If a letter pair repeats (e.g., "LL"), an **X is inserted** to separate them.  
- Encryption rules depend on **letter positions in the matrix**.  

## **Algorithm Steps**  
1. **Generate a 5Ã—5 matrix** from the key (removing duplicate letters and merging J/I).  
2. **Prepare the text**:  
   - Remove non-letters, replace **J â†’ I**, and split into **letter pairs (digraphs)**.  
   - If a pair has duplicate letters, insert **X** between them.  
3. **Encrypt the digraphs**:  
   - **Same row?** Shift each letter **right**.  
   - **Same column?** Shift each letter **down**.  
   - **Rectangle shape?** Swap letters diagonally.  
4. **Decryption** follows the same rules but shifts **left/up instead of right/down**.  

---

## **C++ Implementation**  

```cpp
#include <iostream>
#include <vector>
#include <unordered_set>
using namespace std;

// Generate a 5Ã—5 Playfair Matrix from the key
vector<vector<char>> generateMatrix(string key) {
    string alphabet = "ABCDEFGHIKLMNOPQRSTUVWXYZ"; // J is merged with I
    unordered_set<char> used;
    vector<vector<char>> matrix(5, vector<char>(5));
    string combined = key + alphabet;

    int row = 0, col = 0;
    for (char c : combined) {
        if (used.find(c) == used.end()) {
            matrix[row][col++] = c;
            used.insert(c);
            if (col == 5) {
                col = 0;
                row++;
            }
        }
    }
    return matrix;
}

// Find position of a letter in the matrix
pair<int, int> findPosition(vector<vector<char>> &matrix, char letter) {
    for (int r = 0; r < 5; r++) {
        for (int c = 0; c < 5; c++) {
            if (matrix[r][c] == letter)
                return {r, c};
        }
    }
    return {-1, -1}; // Should never happen
}

// Encrypt or Decrypt using Playfair Cipher
string playfairCipher(string text, vector<vector<char>> &matrix, bool encrypt = true) {
    string processed_text = "";
    
    // Replace J with I & remove non-alphabet chars
    for (char c : text) {
        if (isalpha(c)) processed_text += (toupper(c) == 'J') ? 'I' : toupper(c);
    }

    // Create digraphs (pairs of letters)
    vector<pair<char, char>> digraphs;
    for (size_t i = 0; i < processed_text.length(); i++) {
        char first = processed_text[i];
        char second = (i + 1 < processed_text.length() && processed_text[i] != processed_text[i + 1]) 
                      ? processed_text[++i] : 'X';
        digraphs.push_back({first, second});
    }

    // Encrypt or decrypt each digraph
    string result = "";
    int shift = encrypt ? 1 : -1;

    for (auto [a, b] : digraphs) {
        auto [r1, c1] = findPosition(matrix, a);
        auto [r2, c2] = findPosition(matrix, b);

        if (r1 == r2) { // Same row â†’ Shift right (encrypt) or left (decrypt)
            result += matrix[r1][(c1 + shift + 5) % 5];
            result += matrix[r2][(c2 + shift + 5) % 5];
        } else if (c1 == c2) { // Same column â†’ Shift down (encrypt) or up (decrypt)
            result += matrix[(r1 + shift + 5) % 5][c1];
            result += matrix[(r2 + shift + 5) % 5][c2];
        } else { // Rectangle swap
            result += matrix[r1][c2];
            result += matrix[r2][c1];
        }
    }

    return result;
}

int main() {
    string key, text;
    cout << "Enter key: ";
    cin >> key;

    vector<vector<char>> matrix = generateMatrix(key);

    cout << "Enter text: ";
    cin.ignore();
    getline(cin, text);

    // Encrypt
    string encrypted_text = playfairCipher(text, matrix);
    cout << "Encrypted: " << encrypted_text << endl;

    // Decrypt
    string decrypted_text = playfairCipher(encrypted_text, matrix, false);
    cout << "Decrypted: " << decrypted_text << endl;

    return 0;
}


- **Plaintext:** `"a"`
- **Encrypted Output:** `"DV"`
- **Decrypted Output:** `"AX"`
- **Key (Matrix):** `"CRYPTO"` (used to generate the 5Ã—5 grid)


C R Y P T

O A B D E

F G H I K

L M N Q S

U V W X Z


# 6. **Rail Fence Cipher (Simple Steps)**  

## **Concept**  
- The **Rail Fence Cipher** is a **transposition cipher** that rearranges letters in a zigzag pattern across multiple **rails (rows)**.  
- It does **not change the letters**, just **reorders** them.  
- The number of rails (key) determines the encryption pattern.  

## **Algorithm Steps**  

### **Encryption**  
1. **Create a matrix** with empty rows (rails).  
2. **Write the text in a zigzag pattern** moving up and down between rows.  
3. **Read row-wise** to get the ciphertext.  

### **Decryption**  
1. **Mark positions in the zigzag pattern** based on text length and rails.  
2. **Fill the positions** with ciphertext characters.  
3. **Read in zigzag order** to reconstruct the original text.  

---

## **C++ Implementation**  

```cpp
#include <iostream>
#include <vector>
using namespace std;

// Encrypt using Rail Fence Cipher
string railFenceEncrypt(string text, int key) {
    if (key <= 1) return text; // No encryption needed

    vector<string> rail(key);
    int row = 0, step = 1;

    // Fill the zigzag pattern
    for (char c : text) {
        rail[row] += c;
        if (row == 0) step = 1;
        if (row == key - 1) step = -1;
        row += step;
    }

    // Read row-wise
    string ciphertext;
    for (string r : rail) ciphertext += r;
    return ciphertext;
}

// Decrypt using Rail Fence Cipher
string railFenceDecrypt(string ciphertext, int key) {
    if (key <= 1) return ciphertext; // No decryption needed

    vector<string> rail(key, string(ciphertext.length(), '\n'));
    int row = 0, step = 1;

    // Mark positions for the zigzag pattern
    for (size_t i = 0; i < ciphertext.length(); i++) {
        rail[row][i] = '*';
        if (row == 0) step = 1;
        if (row == key - 1) step = -1;
        row += step;
    }

    // Fill the positions with ciphertext characters
    int index = 0;
    for (int r = 0; r < key; r++) {
        for (size_t c = 0; c < ciphertext.length(); c++) {
            if (rail[r][c] == '*' && index < ciphertext.length()) {
                rail[r][c] = ciphertext[index++];
            }
        }
    }

    // Read in zigzag order to reconstruct the original text
    string plaintext;
    row = 0, step = 1;
    for (size_t i = 0; i < ciphertext.length(); i++) {
        plaintext += rail[row][i];
        if (row == 0) step = 1;
        if (row == key - 1) step = -1;
        row += step;
    }

    return plaintext;
}

int main() {
    string text;
    int rails;

    cout << "Enter text to encrypt: ";
    getline(cin, text);

    cout << "Enter number of rails (key): ";
    cin >> rails;

    // Encrypt
    string encrypted_text = railFenceEncrypt(text, rails);
    cout << "ðŸ” Encrypted: " << encrypted_text << endl;

    // Decrypt
    string decrypted_text = railFenceDecrypt(encrypted_text, rails);
    cout << "ðŸ”“ Decrypted: " << decrypted_text << endl;

    return 0;
}


# 7. **Columnar Transposition Cipher (Simple Steps)**  

## **Concept**  
- The **Columnar Transposition Cipher** rearranges text into a **grid (matrix)** and reads it **column-wise** based on a key.  
- It **shuffles characters** without changing them.  
- The **key determines the column order** during encryption and decryption.  

---

## **Algorithm Steps**  

### **Encryption**  
1. **Write text in a grid row-wise** based on key length.  
2. **Rearrange columns** in the order of the numerical key.  
3. **Read column-wise** to get the ciphertext.  

### **Decryption**  
1. **Recreate the grid** with correct column positions.  
2. **Fill it column-wise** based on the key order.  
3. **Read row-wise** to reconstruct the original text.  

---

## **C++ Implementation**  

```cpp
#include <iostream>
#include <vector>
#include <algorithm>
using namespace std;

// Function to encrypt using Columnar Transposition Cipher
string columnarEncrypt(string text, string key) {
    int numCols = key.length();
    int numRows = (text.length() + numCols - 1) / numCols; // Round up
    vector<vector<char>> grid(numRows, vector<char>(numCols, ' '));

    // Fill grid row-wise
    int index = 0;
    for (int r = 0; r < numRows; r++) {
        for (int c = 0; c < numCols; c++) {
            if (index < text.length()) {
                grid[r][c] = text[index++];
            }
        }
    }

    // Get column order from the key
    vector<pair<int, char>> keyOrder;
    for (int i = 0; i < numCols; i++) {
        keyOrder.push_back({i, key[i]});
    }
    sort(keyOrder.begin(), keyOrder.end(), [](auto &a, auto &b) { return a.second < b.second; });

    // Read column-wise in key order
    string ciphertext;
    for (auto &[c, _] : keyOrder) {
        for (int r = 0; r < numRows; r++) {
            if (grid[r][c] != ' ') {
                ciphertext += grid[r][c];
            }
        }
    }
    return ciphertext;
}

// Function to decrypt using Columnar Transposition Cipher
string columnarDecrypt(string ciphertext, string key) {
    int numCols = key.length();
    int numRows = (ciphertext.length() + numCols - 1) / numCols; // Round up
    vector<vector<char>> grid(numRows, vector<char>(numCols, ' '));

    // Get column order from the key
    vector<pair<int, char>> keyOrder;
    for (int i = 0; i < numCols; i++) {
        keyOrder.push_back({i, key[i]});
    }
    sort(keyOrder.begin(), keyOrder.end(), [](auto &a, auto &b) { return a.second < b.second; });

    // Fill the grid column-wise
    int index = 0;
    for (auto &[c, _] : keyOrder) {
        for (int r = 0; r < numRows; r++) {
            if (index < ciphertext.length()) {
                grid[r][c] = ciphertext[index++];
            }
        }
    }

    // Read row-wise to reconstruct original text
    string plaintext;
    for (int r = 0; r < numRows; r++) {
        for (int c = 0; c < numCols; c++) {
            if (grid[r][c] != ' ') {
                plaintext += grid[r][c];
            }
        }
    }
    return plaintext;
}

int main() {
    string text, key;

    cout << "Enter text to encrypt: ";
    getline(cin, text);

    cout << "Enter numerical key (e.g., 34152): ";
    cin >> key;

    // Encrypt
    string encrypted_text = columnarEncrypt(text, key);
    cout << "ðŸ” Encrypted: " << encrypted_text << endl;

    // Decrypt
    string decrypted_text = columnarDecrypt(encrypted_text, key);
    cout << "ðŸ”“ Decrypted: " << decrypted_text << endl;

    return 0;
}


# 8.**Hill Cipher (Step-by-Step)**  

## **Concept**  
- The **Hill Cipher** is a **block cipher** using **matrix multiplication**.  
- It **encrypts letters in groups** using a **key matrix**.  
- **Decryption requires the modular inverse** of the matrix.  

---

## **Algorithm Steps**  

### **Encryption**  
1. **Convert text to numbers** (A=0, B=1, ..., Z=25).  
2. **Organize numbers into blocks** based on matrix size.  
3. **Multiply each block by the key matrix** (mod 26).  
4. **Convert numbers back to letters** to get the ciphertext.  

### **Decryption**  
1. **Find the modular inverse** of the key matrix.  
2. **Multiply ciphertext blocks by this inverse matrix** (mod 26).  
3. **Convert numbers back to letters** to retrieve the original text.  

---

## **C++ Implementation**  

```cpp
#include <iostream>
#include <vector>
#include <cmath>
using namespace std;

// Function to convert text to numerical representation
vector<int> textToNumbers(string text) {
    vector<int> numbers;
    for (char c : text) {
        if (isalpha(c)) {
            numbers.push_back(toupper(c) - 'A');
        }
    }
    return numbers;
}

// Function to convert numbers back to text
string numbersToText(vector<int> numbers) {
    string text;
    for (int num : numbers) {
        text += (char)(num + 'A');
    }
    return text;
}

// Function to multiply matrix and vector (mod 26)
vector<int> multiplyMatrix(const vector<vector<int>>& matrix, const vector<int>& block) {
    int size = matrix.size();
    vector<int> result(size, 0);
    
    for (int i = 0; i < size; i++) {
        for (int j = 0; j < size; j++) {
            result[i] += matrix[i][j] * block[j];
        }
        result[i] %= 26;
    }
    return result;
}

// Function to compute modular inverse of a number mod 26
int modInverse(int a, int m = 26) {
    a %= m;
    for (int x = 1; x < m; x++) {
        if ((a * x) % m == 1)
            return x;
    }
    return -1;
}

// Function to compute inverse of a 2x2 matrix mod 26
vector<vector<int>> inverseMatrix(const vector<vector<int>>& key) {
    int det = key[0][0] * key[1][1] - key[0][1] * key[1][0];
    det = (det % 26 + 26) % 26;
    int detInv = modInverse(det);
    
    if (detInv == -1) {
        throw runtime_error("Matrix is not invertible under mod 26");
    }

    vector<vector<int>> invMatrix = {
        { key[1][1], -key[0][1] },
        { -key[1][0], key[0][0] }
    };

    for (int i = 0; i < 2; i++) {
        for (int j = 0; j < 2; j++) {
            invMatrix[i][j] = (invMatrix[i][j] * detInv % 26 + 26) % 26;
        }
    }

    return invMatrix;
}

// Function to encrypt text using Hill Cipher
string hillEncrypt(string text, const vector<vector<int>>& key) {
    vector<int> numbers = textToNumbers(text);

    if (numbers.size() % key.size() != 0) {
        numbers.push_back(23); // Padding with 'X' if needed
    }

    vector<int> encryptedNumbers;
    for (size_t i = 0; i < numbers.size(); i += key.size()) {
        vector<int> block(numbers.begin() + i, numbers.begin() + i + key.size());
        vector<int> encryptedBlock = multiplyMatrix(key, block);
        encryptedNumbers.insert(encryptedNumbers.end(), encryptedBlock.begin(), encryptedBlock.end());
    }

    return numbersToText(encryptedNumbers);
}

// Function to decrypt text using Hill Cipher
string hillDecrypt(string text, const vector<vector<int>>& key) {
    vector<int> numbers = textToNumbers(text);
    vector<vector<int>> keyInverse = inverseMatrix(key);

    vector<int> decryptedNumbers;
    for (size_t i = 0; i < numbers.size(); i += keyInverse.size()) {
        vector<int> block(numbers.begin() + i, numbers.begin() + i + keyInverse.size());
        vector<int> decryptedBlock = multiplyMatrix(keyInverse, block);
        decryptedNumbers.insert(decryptedNumbers.end(), decryptedBlock.begin(), decryptedBlock.end());
    }

    return numbersToText(decryptedNumbers);
}

int main() {
    string text;
    vector<vector<int>> key = {{3, 3}, {2, 5}};

    cout << "Enter text to encrypt (A-Z only): ";
    cin >> text;

    // Encrypt
    string encrypted_text = hillEncrypt(text, key);
    cout << "ðŸ” Encrypted: " << encrypted_text << endl;

    // Decrypt
    string decrypted_text = hillDecrypt(encrypted_text, key);
    cout << "ðŸ”“ Decrypted: " << decrypted_text << endl;

    return 0;
}


# 9. **DES (Data Encryption Standard) - Step-by-Step Guide**

## **Concept**
- **DES** is a **symmetric block cipher** that encrypts data in **64-bit blocks**.
- It uses a **56-bit key** (but requires an 8-byte input key).
- Works by applying **multiple rounds of substitution and permutation**.

---

## **Algorithm Steps**

### **Encryption**
1. **Ensure the key is 8 bytes** (DES requires exactly 8 bytes).
2. **Pad the text** to make its length a multiple of 8 bytes.
3. **Use DES in ECB mode** to encrypt the padded text.
4. **Convert the encrypted bytes to Base64** for safe storage.

### **Decryption**
1. **Convert the Base64-encoded text back to bytes**.
2. **Decrypt using the same DES key**.
3. **Remove padding** from the decrypted text.

---

## **C++ Implementation (Crypto++ Library)**

```cpp
#include <iostream>
#include <cryptlib.h>
#include <des.h>
#include <modes.h>
#include <filters.h>
#include <base64.h>

using namespace CryptoPP;
using namespace std;

// Function to pad text to a multiple of 8 bytes
string padText(const string& text) {
    string padded = text;
    while (padded.size() % 8 != 0) {
        padded += " ";  // Padding with spaces
    }
    return padded;
}

// Function to encrypt using DES
string desEncrypt(const string& plaintext, const string& key) {
    string ciphertext;

    ECB_Mode<DES>::Encryption encryptor;
    encryptor.SetKey((byte*)key.data(), key.size());

    StringSource(plaintext, true,
        new StreamTransformationFilter(encryptor,
            new StringSink(ciphertext)
        )
    );

    // Convert to Base64
    string encoded;
    StringSource(ciphertext, true, new Base64Encoder(new StringSink(encoded)));
    return encoded;
}

// Function to decrypt using DES
string desDecrypt(const string& ciphertext, const string& key) {
    string decoded, decrypted;

    // Decode Base64
    StringSource(ciphertext, true, new Base64Decoder(new StringSink(decoded)));

    ECB_Mode<DES>::Decryption decryptor;
    decryptor.SetKey((byte*)key.data(), key.size());

    StringSource(decoded, true,
        new StreamTransformationFilter(decryptor,
            new StringSink(decrypted)
        )
    );

    return decrypted;
}

int main() {
    string plaintext, key;

    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    cout << "Enter an 8-character key: ";
    cin >> key;

    // Ensure key is exactly 8 bytes
    if (key.size() < 8) key.append(8 - key.size(), 'X');
    if (key.size() > 8) key = key.substr(0, 8);

    // Encrypt
    string paddedText = padText(plaintext);
    string encrypted = desEncrypt(paddedText, key);
    cout << "ðŸ” Encrypted: " << encrypted << endl;

    // Decrypt
    string decrypted = desDecrypt(encrypted, key);
    cout << "ðŸ”“ Decrypted: " << decrypted << endl;

    return 0;
}


# 10. **Triple DES (3DES) - Step-by-Step Guide**

## **Concept**
- **Triple DES (3DES)** is an enhancement of the original DES algorithm.
- It **encrypts data three times** using **two or three DES keys** (16 or 24 bytes).
- Uses **multiple rounds of encryption** to improve security.

---

## **Algorithm Steps**

### **Encryption**
1. **Ensure the key is 16 or 24 bytes** (3DES requires exactly this length).
2. **Pad the text** to make its length a multiple of 8 bytes.
3. **Use 3DES in ECB mode** to encrypt the padded text.
4. **Convert the encrypted bytes to Base64** for safe storage.

### **Decryption**
1. **Convert the Base64-encoded text back to bytes**.
2. **Decrypt using the same 3DES key**.
3. **Remove padding** from the decrypted text.

---

## **C++ Implementation (Crypto++ Library)**

```cpp
#include <iostream>
#include <cryptlib.h>
#include <des.h>
#include <modes.h>
#include <filters.h>
#include <base64.h>

using namespace CryptoPP;
using namespace std;

// Function to pad text to a multiple of 8 bytes
string padText(const string& text) {
    string padded = text;
    while (padded.size() % 8 != 0) {
        padded += " ";  // Padding with spaces
    }
    return padded;
}

// Function to encrypt using Triple DES (3DES)
string tripleDesEncrypt(const string& plaintext, const string& key) {
    string ciphertext;

    ECB_Mode<DES_EDE3>::Encryption encryptor;
    encryptor.SetKey((byte*)key.data(), key.size());

    StringSource(plaintext, true,
        new StreamTransformationFilter(encryptor,
            new StringSink(ciphertext)
        )
    );

    // Convert to Base64
    string encoded;
    StringSource(ciphertext, true, new Base64Encoder(new StringSink(encoded)));
    return encoded;
}

// Function to decrypt using Triple DES (3DES)
string tripleDesDecrypt(const string& ciphertext, const string& key) {
    string decoded, decrypted;

    // Decode Base64
    StringSource(ciphertext, true, new Base64Decoder(new StringSink(decoded)));

    ECB_Mode<DES_EDE3>::Decryption decryptor;
    decryptor.SetKey((byte*)key.data(), key.size());

    StringSource(decoded, true,
        new StreamTransformationFilter(decryptor,
            new StringSink(decrypted)
        )
    );

    return decrypted;
}

int main() {
    string plaintext, key;

    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    cout << "Enter a 16 or 24-character key: ";
    cin >> key;

    // Ensure key is exactly 16 or 24 bytes
    if (key.size() < 16) key.append(16 - key.size(), 'X');
    if (key.size() > 24) key = key.substr(0, 24);

    // Encrypt
    string paddedText = padText(plaintext);
    string encrypted = tripleDesEncrypt(paddedText, key);
    cout << "ðŸ” Encrypted: " << encrypted << endl;

    // Decrypt
    string decrypted = tripleDesDecrypt(encrypted, key);
    cout << "ðŸ”“ Decrypted: " << decrypted << endl;

    return 0;
}


# 11. **AES Encryption (ECB Mode) - Step-by-Step Guide**

## **Concept**
- **AES (Advanced Encryption Standard)** is a symmetric key encryption algorithm.
- Encrypts data in **fixed 128-bit (16-byte) blocks**.
- Uses a **secret key** (must be shared between sender & receiver).

---

## **Algorithm Steps**

### **Encryption**
1. **Generate a 16-byte (128-bit) key** (randomly or user-defined).
2. **Pad the plaintext** to make its length a multiple of 16 bytes.
3. **Encrypt using AES in ECB mode**.
4. **Convert encrypted bytes to Base64** for safe storage.

### **Decryption**
1. **Convert the Base64-encoded text back to bytes**.
2. **Decrypt using the same AES key**.
3. **Remove padding** from the decrypted text.

---

## **C++ Implementation (Crypto++ Library)**

```cpp
#include <iostream>
#include <cryptlib.h>
#include <aes.h>
#include <modes.h>
#include <filters.h>
#include <base64.h>
#include <osrng.h>

using namespace CryptoPP;
using namespace std;

// Function to pad text to a multiple of 16 bytes
string padText(const string& text) {
    string padded = text;
    size_t paddingSize = 16 - (padded.size() % 16);
    padded.append(paddingSize, paddingSize);  // Padding with bytes of same value
    return padded;
}

// Function to remove padding after decryption
string unpadText(const string& text) {
    return text.substr(0, text.size() - text.back());
}

// Function to generate a random 16-byte key
string generateKey() {
    AutoSeededRandomPool prng;
    SecByteBlock key(AES::DEFAULT_KEYLENGTH);
    prng.GenerateBlock(key, key.size());
    return string((const char*)key.data(), key.size());
}

// Function to encrypt using AES (ECB mode)
string aesEncrypt(const string& plaintext, const string& key) {
    string ciphertext;

    ECB_Mode<AES>::Encryption encryptor;
    encryptor.SetKey((byte*)key.data(), key.size());

    StringSource(padText(plaintext), true,
        new StreamTransformationFilter(encryptor,
            new StringSink(ciphertext)
        )
    );

    // Convert to Base64
    string encoded;
    StringSource(ciphertext, true, new Base64Encoder(new StringSink(encoded)));
    return encoded;
}

// Function to decrypt using AES (ECB mode)
string aesDecrypt(const string& ciphertext, const string& key) {
    string decoded, decrypted;

    // Decode Base64
    StringSource(ciphertext, true, new Base64Decoder(new StringSink(decoded)));

    ECB_Mode<AES>::Decryption decryptor;
    decryptor.SetKey((byte*)key.data(), key.size());

    StringSource(decoded, true,
        new StreamTransformationFilter(decryptor,
            new StringSink(decrypted)
        )
    );

    return unpadText(decrypted);
}

int main() {
    string plaintext;
    
    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    // Generate a random AES key
    string key = generateKey();

    // Encrypt
    string encrypted = aesEncrypt(plaintext, key);
    cout << "ðŸ” Encrypted: " << encrypted << endl;

    // Decrypt
    string decrypted = aesDecrypt(encrypted, key);
    cout << "ðŸ”“ Decrypted: " << decrypted << endl;

    return 0;
}


# 12. **Blowfish Encryption (ECB Mode) - Step-by-Step Guide**

## **Concept**
- **Blowfish** is a symmetric key block cipher with **variable key size (4 to 56 bytes)**.
- Encrypts data in **64-bit (8-byte) blocks**.
- Uses a **secret key** for both encryption and decryption.

---

## **Algorithm Steps**

### **Encryption**
1. **Generate a key** (4-56 bytes).
2. **Pad the plaintext** to make its length a multiple of 8 bytes.
3. **Encrypt using Blowfish in ECB mode**.
4. **Convert encrypted bytes to Base64** for safe storage.

### **Decryption**
1. **Convert the Base64-encoded text back to bytes**.
2. **Decrypt using the same Blowfish key**.
3. **Remove padding** from the decrypted text.

---

## **C++ Implementation (Crypto++ Library)**

```cpp
#include <iostream>
#include <cryptlib.h>
#include <blowfish.h>
#include <modes.h>
#include <filters.h>
#include <base64.h>
#include <osrng.h>

using namespace CryptoPP;
using namespace std;

// Function to pad text to a multiple of 8 bytes
string padText(const string& text) {
    string padded = text;
    size_t paddingSize = 8 - (padded.size() % 8);
    padded.append(paddingSize, paddingSize);  // Padding with bytes of same value
    return padded;
}

// Function to remove padding after decryption
string unpadText(const string& text) {
    return text.substr(0, text.size() - text.back());
}

// Function to encrypt using Blowfish (ECB mode)
string blowfishEncrypt(const string& plaintext, const string& key) {
    string ciphertext;

    ECB_Mode<Blowfish>::Encryption encryptor;
    encryptor.SetKey((byte*)key.data(), key.size());

    StringSource(padText(plaintext), true,
        new StreamTransformationFilter(encryptor,
            new StringSink(ciphertext)
        )
    );

    // Convert to Base64
    string encoded;
    StringSource(ciphertext, true, new Base64Encoder(new StringSink(encoded)));
    return encoded;
}

// Function to decrypt using Blowfish (ECB mode)
string blowfishDecrypt(const string& ciphertext, const string& key) {
    string decoded, decrypted;

    // Decode Base64
    StringSource(ciphertext, true, new Base64Decoder(new StringSink(decoded)));

    ECB_Mode<Blowfish>::Decryption decryptor;
    decryptor.SetKey((byte*)key.data(), key.size());

    StringSource(decoded, true,
        new StreamTransformationFilter(decryptor,
            new StringSink(decrypted)
        )
    );

    return unpadText(decrypted);
}

int main() {
    string plaintext, key;
    
    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    cout << "Enter Blowfish key (4-56 bytes): ";
    getline(cin, key);

    // Encrypt
    string encrypted = blowfishEncrypt(plaintext, key);
    cout << "ðŸ” Encrypted: " << encrypted << endl;

    // Decrypt
    string decrypted = blowfishDecrypt(encrypted, key);
    cout << "ðŸ”“ Decrypted: " << decrypted << endl;

    return 0;
}


# **Blowfish Encryption (CBC Mode) - Step-by-Step Guide**

## **Concept**
- **Blowfish** is a **symmetric key block cipher** with a **variable key size (4 to 56 bytes)**.
- Encrypts data in **64-bit (8-byte) blocks**.
- Uses a **secret key** for both encryption and decryption.
- **CBC (Cipher Block Chaining) Mode** introduces **IV (Initialization Vector)** for added security.

---

## **Algorithm Steps**

### **Encryption**
1. **Generate a key** (4-56 bytes).
2. **Generate a random 8-byte IV (Initialization Vector)**.
3. **Pad the plaintext** to make its length a multiple of 8 bytes.
4. **Encrypt using Blowfish in CBC mode**.
5. **Concatenate IV with the ciphertext and encode in Base64** for storage.

### **Decryption**
1. **Decode Base64-encoded text back to bytes**.
2. **Extract the IV from the first 8 bytes**.
3. **Decrypt using the same Blowfish key and IV**.
4. **Remove padding** from the decrypted text.

---

## **C++ Implementation (Crypto++ Library)**

```cpp
#include <iostream>
#include <cryptlib.h>
#include <blowfish.h>
#include <modes.h>
#include <filters.h>
#include <base64.h>
#include <osrng.h>

using namespace CryptoPP;
using namespace std;

// Function to generate a random IV (8 bytes)
string generateIV() {
    AutoSeededRandomPool rng;
    byte iv[Blowfish::BLOCKSIZE];
    rng.GenerateBlock(iv, sizeof(iv));
    return string((char*)iv, Blowfish::BLOCKSIZE);
}

// Function to encrypt using Blowfish (CBC mode)
string blowfishEncrypt(const string& plaintext, const string& key, string& iv) {
    string ciphertext;
    CBC_Mode<Blowfish>::Encryption encryptor;
    
    encryptor.SetKeyWithIV((byte*)key.data(), key.size(), (byte*)iv.data());

    StringSource(plaintext, true,
        new StreamTransformationFilter(encryptor,
            new StringSink(ciphertext)
        )
    );

    return iv + ciphertext;  // Store IV + ciphertext together
}

// Function to decrypt using Blowfish (CBC mode)
string blowfishDecrypt(const string& encrypted, const string& key) {
    string iv = encrypted.substr(0, Blowfish::BLOCKSIZE);  // Extract IV
    string ciphertext = encrypted.substr(Blowfish::BLOCKSIZE);  // Extract actual ciphertext
    string decrypted;

    CBC_Mode<Blowfish>::Decryption decryptor;
    decryptor.SetKeyWithIV((byte*)key.data(), key.size(), (byte*)iv.data());

    StringSource(ciphertext, true,
        new StreamTransformationFilter(decryptor,
            new StringSink(decrypted)
        )
    );

    return decrypted;
}

int main() {
    string plaintext, key;
    
    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    cout << "Enter Blowfish key (4-56 bytes): ";
    getline(cin, key);

    // Generate IV
    string iv = generateIV();

    // Encrypt
    string encrypted = blowfishEncrypt(plaintext, key, iv);
    string encoded;
    StringSource(encrypted, true, new Base64Encoder(new StringSink(encoded)));
    cout << "ðŸ” Encrypted (Base64): " << encoded << endl;

    // Decode Base64 before decryption
    string decoded;
    StringSource(encoded, true, new Base64Decoder(new StringSink(decoded)));

    // Decrypt
    string decrypted = blowfishDecrypt(decoded, key);
    cout << "ðŸ”“ Decrypted: " << decrypted << endl;

    return 0;
}


# Twofish (env. error)

# **Twofish Encryption (Step-by-Step Guide)**

## **Concept**
- **Twofish** is a **symmetric key block cipher** and a successor of Blowfish.
- Encrypts data in **128-bit (16-byte) blocks**.
- Supports **key sizes of 16, 24, or 32 bytes**.
- Faster and more secure than Blowfish.

---

## **Algorithm Steps**

### **Encryption**
1. **Generate a key** (16, 24, or 32 bytes).
2. **Pad the plaintext** to 16 bytes if necessary.
3. **Encrypt using Twofish**.
4. **Encode the ciphertext in Base64** for storage.

### **Decryption**
1. **Decode Base64-encoded text back to bytes**.
2. **Decrypt using the same Twofish key**.
3. **Remove padding from the decrypted text**.

---

## **C++ Implementation (Crypto++ Library)**

```cpp
#include <iostream>
#include <cryptlib.h>
#include <twofish.h>
#include <modes.h>
#include <filters.h>
#include <base64.h>

using namespace CryptoPP;
using namespace std;

// Function to encrypt using Twofish (ECB Mode)
string twofishEncrypt(const string& plaintext, const string& key) {
    string ciphertext;

    ECB_Mode<Twofish>::Encryption encryptor;
    encryptor.SetKey((byte*)key.data(), key.size());

    StringSource(plaintext, true,
        new StreamTransformationFilter(encryptor,
            new StringSink(ciphertext)
        )
    );

    return ciphertext;
}

// Function to decrypt using Twofish (ECB Mode)
string twofishDecrypt(const string& encrypted, const string& key) {
    string decrypted;

    ECB_Mode<Twofish>::Decryption decryptor;
    decryptor.SetKey((byte*)key.data(), key.size());

    StringSource(encrypted, true,
        new StreamTransformationFilter(decryptor,
            new StringSink(decrypted)
        )
    );

    return decrypted;
}

int main() {
    string plaintext, key;

    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    cout << "Enter Twofish key (16, 24, or 32 bytes): ";
    getline(cin, key);

    // Pad key if needed
    if (key.size() < 16) key.append(16 - key.size(), 'X');
    key = key.substr(0, 16); // Ensure key is exactly 16 bytes

    // Encrypt
    string encrypted = twofishEncrypt(plaintext, key);
    string encoded;
    StringSource(encrypted, true, new Base64Encoder(new StringSink(encoded)));
    cout << "ðŸ” Encrypted (Base64): " << encoded << endl;

    // Decode Base64 before decryption
    string decoded;
    StringSource(encoded, true, new Base64Decoder(new StringSink(decoded)));

    // Decrypt
    string decrypted = twofishDecrypt(decoded, key);
    cout << "ðŸ”“ Decrypted: " << decrypted << endl;

    return 0;
}


# 13. **IDEA (International Data Encryption Algorithm)**
## **Concept**
- **IDEA** is a **symmetric-key block cipher**.
- Uses a **64-bit block size**.
- Requires a **128-bit (16-byte) key**.
- Consists of **8 encryption rounds** + 1 final transformation.
- Operates in different modes like **ECB, CBC, CFB, OFB**.

---

## **Algorithm Steps**
### **Encryption**
1. **Generate a 16-byte key** (128-bit).
2. **Choose an 8-byte IV** (for CBC mode).
3. **Pad plaintext** to be a multiple of 8 bytes.
4. **Encrypt using IDEA**.
5. **Store IV + ciphertext (Base64 encoded).**

### **Decryption**
1. **Extract IV from the encrypted message**.
2. **Decrypt using the same IDEA key**.
3. **Remove padding** to get the original text.

---

## **C++ Implementation (Crypto++ Library)**

```cpp
#include <iostream>
#include <cryptlib.h>
#include <idea.h>
#include <modes.h>
#include <filters.h>
#include <base64.h>

using namespace CryptoPP;
using namespace std;

// Function to encrypt using IDEA (CBC Mode)
string ideaEncrypt(const string& plaintext, const string& key) {
    string ciphertext, encoded;
    SecByteBlock keyBytes((const byte*)key.data(), key.size());
    byte iv[IDEA::BLOCKSIZE];
    OS_GenerateRandomBlock(false, iv, IDEA::BLOCKSIZE);

    CBC_Mode<IDEA>::Encryption encryptor;
    encryptor.SetKeyWithIV(keyBytes, keyBytes.size(), iv);

    StringSource(plaintext, true,
        new StreamTransformationFilter(encryptor,
            new StringSink(ciphertext)
        )
    );

    // Encode IV + Ciphertext to Base64
    string ivStr(reinterpret_cast<const char*>(iv), IDEA::BLOCKSIZE);
    StringSource(ivStr + ciphertext, true, new Base64Encoder(new StringSink(encoded)));
    return encoded;
}

// Function to decrypt using IDEA (CBC Mode)
string ideaDecrypt(const string& encrypted, const string& key) {
    string decoded, decrypted;
    SecByteBlock keyBytes((const byte*)key.data(), key.size());
    byte iv[IDEA::BLOCKSIZE];

    // Decode Base64
    StringSource(encrypted, true, new Base64Decoder(new StringSink(decoded)));

    // Extract IV
    memcpy(iv, decoded.data(), IDEA::BLOCKSIZE);
    string ciphertext = decoded.substr(IDEA::BLOCKSIZE);

    CBC_Mode<IDEA>::Decryption decryptor;
    decryptor.SetKeyWithIV(keyBytes, keyBytes.size(), iv);

    StringSource(ciphertext, true,
        new StreamTransformationFilter(decryptor,
            new StringSink(decrypted)
        )
    );

    return decrypted;
}

int main() {
    string plaintext, key;

    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    cout << "Enter IDEA key (16 bytes): ";
    getline(cin, key);

    // Ensure key is exactly 16 bytes
    if (key.size() < 16) key.append(16 - key.size(), 'X');
    key = key.substr(0, 16);

    // Encrypt
    string encrypted = ideaEncrypt(plaintext, key);
    cout << "ðŸ” Encrypted (Base64): " << encrypted << endl;

    // Decrypt
    string decrypted = ideaDecrypt(encrypted, key);
    cout << "ðŸ”“ Decrypted: " << decrypted << endl;

    return 0;
}


# 14. **RC6 (Rivest Cipher 6)**
## **Concept**
- **RC6** is a **symmetric-key block cipher**.
- Designed as an improvement over **RC5**.
- Uses a **128-bit block size** (16 bytes).
- Supports **128, 192, or 256-bit keys**.
- Operates in **20 rounds** (default).

---

## **Algorithm Steps**
### **Key Schedule**
1. Convert the key into **32-bit words**.
2. Generate **subkeys** using constants **P32** and **Q32**.
3. Perform **mixing rounds** with bit rotations.

### **Encryption**
1. Split 128-bit block into **A, B, C, D** (32-bit each).
2. Add **S[0]** to **B** and **S[1]** to **D**.
3. Iterate **20 rounds**, applying:
   - **Bitwise XOR**
   - **Multiplication**
   - **Left rotations**
   - **Key mixing**
4. Final transformations add **S[42]** to **A** and **S[43]** to **C**.

### **Decryption**
1. Reverse the process with **inverse rotations** and **subkey subtraction**.

---

## **C++ Implementation (Crypto++ Library)**

```cpp
#include <iostream>
#include <cryptlib.h>
#include <rc6.h>
#include <modes.h>
#include <filters.h>
#include <hex.h>

using namespace CryptoPP;
using namespace std;

// Function to encrypt using RC6
string rc6Encrypt(const string& plaintext, const string& key) {
    string ciphertext, encoded;
    SecByteBlock keyBytes((const byte*)key.data(), key.size());

    ECB_Mode<RC6>::Encryption encryptor;
    encryptor.SetKey(keyBytes, keyBytes.size());

    StringSource(plaintext, true,
        new StreamTransformationFilter(encryptor,
            new StringSink(ciphertext)
        )
    );

    // Convert ciphertext to hex format
    StringSource(ciphertext, true, new HexEncoder(new StringSink(encoded)));
    return encoded;
}

// Function to decrypt using RC6
string rc6Decrypt(const string& encrypted, const string& key) {
    string decoded, decrypted;
    SecByteBlock keyBytes((const byte*)key.data(), key.size());

    ECB_Mode<RC6>::Decryption decryptor;
    decryptor.SetKey(keyBytes, keyBytes.size());

    // Convert hex back to binary
    StringSource(encrypted, true, new HexDecoder(new StringSink(decoded)));

    StringSource(decoded, true,
        new StreamTransformationFilter(decryptor,
            new StringSink(decrypted)
        )
    );

    return decrypted;
}

int main() {
    string plaintext, key;

    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    cout << "Enter RC6 key (16, 24, or 32 bytes): ";
    getline(cin, key);

    // Ensure key is 16, 24, or 32 bytes
    if (key.size() < 16) key.append(16 - key.size(), 'X');
    key = key.substr(0, 16);

    // Encrypt
    string encrypted = rc6Encrypt(plaintext, key);
    cout << "ðŸ” Encrypted (Hex): " << encrypted << endl;

    // Decrypt
    string decrypted = rc6Decrypt(encrypted, key);
    cout << "ðŸ”“ Decrypted: " << decrypted << endl;

    return 0;
}


# 15. **RC4 (Rivest Cipher 4)**
## **Concept**
- **RC4** is a **stream cipher** (not a block cipher).
- Uses a **variable-length key (typically 5â€“16 bytes)**.
- Generates a **pseudo-random keystream** to XOR with plaintext.
- Simple but **insecure** due to known vulnerabilities (not recommended for secure use).

---

## **Algorithm Steps**
### **Key Scheduling Algorithm (KSA)**
1. Initialize a **256-byte state array (S)**.
2. Populate S with values **0â€“255**.
3. Perform **key-dependent permutations** on S.

### **Pseudo-Random Generation Algorithm (PRGA)**
1. Generate a **keystream byte-by-byte**.
2. XOR the keystream with the plaintext to **encrypt**.
3. The same process **decrypts** the ciphertext.

---

## **C++ Implementation (Crypto++ Library)**

```cpp
#include <iostream>
#include <cryptlib.h>
#include <rc4.h>
#include <hex.h>
#include <filters.h>

using namespace CryptoPP;
using namespace std;

// Function to encrypt using RC4
string rc4Encrypt(const string& plaintext, const string& key) {
    string ciphertext, encoded;
    SecByteBlock keyBytes((const byte*)key.data(), key.size());

    RC4::Encryption encryptor;
    encryptor.SetKey(keyBytes, keyBytes.size());

    StringSource(plaintext, true,
        new StreamTransformationFilter(encryptor,
            new StringSink(ciphertext)
        )
    );

    // Convert to hex format for easier display
    StringSource(ciphertext, true, new HexEncoder(new StringSink(encoded)));
    return encoded;
}

// Function to decrypt using RC4
string rc4Decrypt(const string& encrypted, const string& key) {
    string decoded, decrypted;
    SecByteBlock keyBytes((const byte*)key.data(), key.size());

    RC4::Decryption decryptor;
    decryptor.SetKey(keyBytes, keyBytes.size());

    // Convert hex back to binary
    StringSource(encrypted, true, new HexDecoder(new StringSink(decoded)));

    StringSource(decoded, true,
        new StreamTransformationFilter(decryptor,
            new StringSink(decrypted)
        )
    );

    return decrypted;
}

int main() {
    string plaintext, key;

    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    cout << "Enter RC4 key (5-16 bytes recommended): ";
    getline(cin, key);

    // Ensure key is at least 5 bytes
    if (key.size() < 5) key.append(5 - key.size(), 'X');

    // Encrypt
    string encrypted = rc4Encrypt(plaintext, key);
    cout << "ðŸ” Encrypted (Hex): " << encrypted << endl;

    // Decrypt
    string decrypted = rc4Decrypt(encrypted, key);
    cout << "ðŸ”“ Decrypted: " << decrypted << endl;

    return 0;
}


# 16. **ChaCha20 Cipher**
## **Concept**
- **ChaCha20** is a **stream cipher** designed as a secure alternative to RC4.
- It uses a **256-bit key** and a **96-bit nonce** (some implementations use 64-bit nonces).
- Highly resistant to cryptanalysis and widely used in **TLS, SSH, and modern encryption**.

---

## **Algorithm Steps**
1. Initialize ChaCha20 with:
   - **256-bit key (32 bytes)**
   - **Nonce (64-bit or 96-bit)**
2. Generate a **keystream** using a pseudo-random function.
3. XOR keystream with plaintext to produce ciphertext.
4. For decryption, XOR ciphertext with the same keystream.

---

## **C++ Implementation (Crypto++)**
```cpp
#include <iostream>
#include <cryptlib.h>
#include <chacha.h>
#include <hex.h>
#include <osrng.h>
#include <filters.h>

using namespace CryptoPP;
using namespace std;

// Function to encrypt using ChaCha20
string chacha20Encrypt(const string& plaintext, const SecByteBlock& key, const byte nonce[8]) {
    string ciphertext, encoded;

    ChaCha20::Encryption encryptor;
    encryptor.SetKeyWithIV(key, key.size(), nonce, 8);

    StringSource(plaintext, true,
        new StreamTransformationFilter(encryptor,
            new StringSink(ciphertext)
        )
    );

    // Convert to hex for display
    StringSource(ciphertext, true, new HexEncoder(new StringSink(encoded)));
    return encoded;
}

// Function to decrypt using ChaCha20
string chacha20Decrypt(const string& encrypted, const SecByteBlock& key, const byte nonce[8]) {
    string decoded, decrypted;

    // Convert hex back to binary
    StringSource(encrypted, true, new HexDecoder(new StringSink(decoded)));

    ChaCha20::Decryption decryptor;
    decryptor.SetKeyWithIV(key, key.size(), nonce, 8);

    StringSource(decoded, true,
        new StreamTransformationFilter(decryptor,
            new StringSink(decrypted)
        )
    );

    return decrypted;
}

int main() {
    AutoSeededRandomPool prng;

    SecByteBlock key(ChaCha20::DEFAULT_KEYLENGTH);  // 256-bit key
    prng.GenerateBlock(key, key.size());

    byte nonce[8];  // 64-bit nonce
    prng.GenerateBlock(nonce, sizeof(nonce));

    string plaintext;
    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    // Encrypt
    string encrypted = chacha20Encrypt(plaintext, key, nonce);
    cout << "ðŸ” Encrypted (Hex): " << encrypted << endl;

    // Decrypt
    string decrypted = chacha20Decrypt(encrypted, key, nonce);
    cout << "ðŸ”“ Decrypted: " << decrypted << endl;

    return 0;
}


# 16(a). **ChaCha20-Poly1305 Cipher**
## **Concept**
- **ChaCha20-Poly1305** is an **AEAD (Authenticated Encryption with Associated Data)** cipher.
- Combines **ChaCha20** (for encryption) and **Poly1305** (for authentication).
- Used in **TLS 1.3, SSH, and modern encryption protocols**.

---

## **Algorithm Steps**
1. Initialize ChaCha20-Poly1305 with:
   - **256-bit key (32 bytes)**
   - **96-bit nonce (12 bytes)**
   - **Optional associated data (AD)**
2. Encrypt plaintext using **ChaCha20** stream cipher.
3. Generate **authentication tag** using **Poly1305**.
4. For decryption:
   - Verify **authentication tag** before decrypting.
   - If verification fails, decryption is **rejected**.

---

## **C++ Implementation (Crypto++)**
```cpp
#include <iostream>
#include <cryptlib.h>
#include <chachapoly.h>
#include <hex.h>
#include <osrng.h>
#include <filters.h>

using namespace CryptoPP;
using namespace std;

// Function to encrypt using ChaCha20-Poly1305
string chacha20Poly1305Encrypt(const string& plaintext, const SecByteBlock& key, const byte nonce[12], const string& associatedData) {
    string ciphertext, encoded, tag;

    ChaCha20Poly1305::Encryption encryptor;
    encryptor.SetKeyWithIV(key, key.size(), nonce, 12);
    encryptor.SpecifyDataLengths(associatedData.size(), plaintext.size(), 0);

    AuthenticatedEncryptionFilter encryptFilter(encryptor,
        new StringSink(ciphertext), false, 16 // 16-byte tag
    );
    
    encryptFilter.ChannelPut(AAD_CHANNEL, (const byte*)associatedData.data(), associatedData.size());
    encryptFilter.ChannelMessageEnd(AAD_CHANNEL);

    encryptFilter.ChannelPut(DEFAULT_CHANNEL, (const byte*)plaintext.data(), plaintext.size());
    encryptFilter.ChannelMessageEnd(DEFAULT_CHANNEL);

    return ciphertext;
}

// Function to decrypt using ChaCha20-Poly1305
string chacha20Poly1305Decrypt(const string& encrypted, const SecByteBlock& key, const byte nonce[12], const string& associatedData) {
    string decrypted;
    
    ChaCha20Poly1305::Decryption decryptor;
    decryptor.SetKeyWithIV(key, key.size(), nonce, 12);
    decryptor.SpecifyDataLengths(associatedData.size(), encrypted.size() - 16, 16);  // Subtract tag size

    AuthenticatedDecryptionFilter decryptFilter(decryptor,
        new StringSink(decrypted), AuthenticatedDecryptionFilter::MAC_AT_END, 16
    );

    decryptFilter.ChannelPut(AAD_CHANNEL, (const byte*)associatedData.data(), associatedData.size());
    decryptFilter.ChannelMessageEnd(AAD_CHANNEL);

    decryptFilter.ChannelPut(DEFAULT_CHANNEL, (const byte*)encrypted.data(), encrypted.size());
    decryptFilter.ChannelMessageEnd(DEFAULT_CHANNEL);

    if (!decryptFilter.GetLastResult()) {
        throw runtime_error("ðŸ”´ Authentication failed!");
    }

    return decrypted;
}

int main() {
    AutoSeededRandomPool prng;

    SecByteBlock key(ChaCha20Poly1305::DEFAULT_KEYLENGTH);  // 256-bit key
    prng.GenerateBlock(key, key.size());

    byte nonce[12];  // 96-bit nonce
    prng.GenerateBlock(nonce, sizeof(nonce));

    string plaintext, associatedData = "AUTH_DATA";
    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    // Encrypt
    string encrypted = chacha20Poly1305Encrypt(plaintext, key, nonce, associatedData);
    cout << "ðŸ” Encrypted (Hex): " << HexEncoder(new StringSink(encrypted), true).Ref() << endl;

    // Decrypt
    try {
        string decrypted = chacha20Poly1305Decrypt(encrypted, key, nonce, associatedData);
        cout << "ðŸ”“ Decrypted: " << decrypted << endl;
    } catch (const exception& e) {
        cout << e.what() << endl;
    }

    return 0;
}


# 16(b). **XChaCha20-Poly1305 Cipher**
## **Concept**
- **XChaCha20-Poly1305** is an **AEAD (Authenticated Encryption with Associated Data)** cipher.
- Uses **XChaCha20** (an extended nonce version of ChaCha20) for encryption and **Poly1305** for authentication.
- Extended **192-bit nonce (24 bytes)** prevents nonce reuse, making it ideal for **databases, TLS, and secure messaging**.

---

## **Algorithm Steps**
1. Generate a **random 192-bit nonce (24 bytes)**.
2. Initialize **XChaCha20-Poly1305** with:
   - **256-bit key (32 bytes)**
   - **24-byte nonce**
   - **Optional associated data (AD)**
3. Encrypt plaintext using **XChaCha20** stream cipher.
4. Generate **authentication tag** using **Poly1305**.
5. For decryption:
   - Verify **authentication tag** before decrypting.
   - If verification fails, decryption is **rejected**.

---

## **C++ Implementation (libsodium)**
```cpp
#include <sodium.h>
#include <iostream>
#include <string>
#include <vector>

using namespace std;

// Function to encrypt using XChaCha20-Poly1305
vector<unsigned char> xchacha20_poly1305_encrypt(const string& plaintext, const vector<unsigned char>& key, const string& associatedData) {
    vector<unsigned char> nonce(crypto_aead_xchacha20poly1305_ietf_NPUBBYTES);
    randombytes_buf(nonce.data(), nonce.size()); // Generate random nonce

    vector<unsigned char> ciphertext(plaintext.size() + crypto_aead_xchacha20poly1305_ietf_ABYTES);
    unsigned long long ciphertext_len;

    crypto_aead_xchacha20poly1305_ietf_encrypt(
        ciphertext.data(), &ciphertext_len,
        reinterpret_cast<const unsigned char*>(plaintext.data()), plaintext.size(),
        reinterpret_cast<const unsigned char*>(associatedData.data()), associatedData.size(),
        NULL, nonce.data(), key.data()
    );

    // Store nonce + ciphertext
    ciphertext.insert(ciphertext.begin(), nonce.begin(), nonce.end());
    return ciphertext;
}

// Function to decrypt using XChaCha20-Poly1305
string xchacha20_poly1305_decrypt(const vector<unsigned char>& encrypted, const vector<unsigned char>& key, const string& associatedData) {
    vector<unsigned char> nonce(encrypted.begin(), encrypted.begin() + crypto_aead_xchacha20poly1305_ietf_NPUBBYTES);
    vector<unsigned char> ciphertext(encrypted.begin() + crypto_aead_xchacha20poly1305_ietf_NPUBBYTES, encrypted.end());

    vector<unsigned char> decrypted(ciphertext.size() - crypto_aead_xchacha20poly1305_ietf_ABYTES);
    unsigned long long decrypted_len;

    if (crypto_aead_xchacha20poly1305_ietf_decrypt(
            decrypted.data(), &decrypted_len, NULL,
            ciphertext.data(), ciphertext.size(),
            reinterpret_cast<const unsigned char*>(associatedData.data()), associatedData.size(),
            nonce.data(), key.data()) != 0) {
        throw runtime_error("ðŸ”´ Authentication failed!");
    }

    return string(decrypted.begin(), decrypted.begin() + decrypted_len);
}

int main() {
    if (sodium_init() < 0) {
        cerr << "Failed to initialize libsodium!" << endl;
        return 1;
    }

    vector<unsigned char> key(crypto_aead_xchacha20poly1305_ietf_KEYBYTES);
    randombytes_buf(key.data(), key.size()); // Generate 256-bit key

    string plaintext, associatedData = "AUTH_DATA";
    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    // Encrypt
    vector<unsigned char> encrypted = xchacha20_poly1305_encrypt(plaintext, key, associatedData);
    cout << "ðŸ” Encrypted (Hex): ";
    for (unsigned char byte : encrypted) {
        printf("%02x", byte);
    }
    cout << endl;

    // Decrypt
    try {
        string decrypted = xchacha20_poly1305_decrypt(encrypted, key, associatedData);
        cout << "ðŸ”“ Decrypted: " << decrypted << endl;
    } catch (const exception& e) {
        cout << e.what() << endl;
    }

    return 0;
}


# 17. **Salsa20 Cipher**
## **Concept**
- **Salsa20** is a **stream cipher** designed by **Daniel J. Bernstein**.
- It provides high-speed encryption with **256-bit keys (32 bytes)**.
- Uses a **64-bit nonce (8 bytes)** to ensure unique encryption instances.
- A variant, **XSalsa20**, extends the nonce to **192 bits** for enhanced security.

---

## **Algorithm Steps**
1. Generate a **random 64-bit nonce (8 bytes)**.
2. Initialize **Salsa20** with:
   - **256-bit key (32 bytes)**
   - **8-byte nonce**
3. Encrypt plaintext using **Salsa20 keystream**.
4. For decryption:
   - Use the **same nonce** and **key** to regenerate the keystream.
   - XOR ciphertext with keystream to retrieve plaintext.

---

## **C++ Implementation (Crypto++)**
```cpp
#include <iostream>
#include <cryptlib.h>
#include <salsa.h>
#include <hex.h>
#include <osrng.h>
#include <filters.h>

using namespace CryptoPP;
using namespace std;

// Function to encrypt using Salsa20
string salsa20Encrypt(const string& plaintext, const SecByteBlock& key, const byte nonce[8]) {
    string ciphertext;

    Salsa20::Encryption salsa;
    salsa.SetKeyWithIV(key, key.size(), nonce, 8); // Set key and nonce

    StringSource(plaintext, true,
        new StreamTransformationFilter(salsa,
            new StringSink(ciphertext)
        )
    );

    return ciphertext;
}

// Function to decrypt using Salsa20
string salsa20Decrypt(const string& encrypted, const SecByteBlock& key, const byte nonce[8]) {
    string decrypted;

    Salsa20::Decryption salsa;
    salsa.SetKeyWithIV(key, key.size(), nonce, 8); // Set key and nonce

    StringSource(encrypted, true,
        new StreamTransformationFilter(salsa,
            new StringSink(decrypted)
        )
    );

    return decrypted;
}

int main() {
    AutoSeededRandomPool prng;

    SecByteBlock key(Salsa20::DEFAULT_KEYLENGTH); // 256-bit key
    prng.GenerateBlock(key, key.size());

    byte nonce[8]; // 64-bit nonce
    prng.GenerateBlock(nonce, sizeof(nonce));

    string plaintext;
    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    // Encrypt
    string encrypted = salsa20Encrypt(plaintext, key, nonce);
    cout << "ðŸ” Encrypted (Hex): ";
    StringSource(encrypted, true, new HexEncoder(new FileSink(cout)));
    cout << endl;

    // Decrypt
    string decrypted = salsa20Decrypt(encrypted, key, nonce);
    cout << "ðŸ”“ Decrypted: " << decrypted << endl;

    return 0;
}


# 17(a). **Salsa20-Poly1305 Cipher**
## **Concept**
- **Salsa20-Poly1305** is an **AEAD (Authenticated Encryption with Associated Data)** cipher.
- Combines **Salsa20** (for encryption) and **Poly1305** (for authentication).
- Used in **secure messaging, VPNs, and real-time encryption protocols**.

---

## **Algorithm Steps**
1. Generate a **random 64-bit nonce (8 bytes)**.
2. Initialize **Salsa20-Poly1305** with:
   - **256-bit key (32 bytes)**
   - **8-byte nonce**
   - **Optional associated data (AD)**
3. Encrypt plaintext using **Salsa20** stream cipher.
4. Generate **authentication tag** using **Poly1305**.
5. For decryption:
   - Verify **authentication tag** before decrypting.
   - If verification fails, decryption is **rejected**.

---

## **C++ Implementation (Crypto++)**
```cpp
#include <iostream>
#include <cryptlib.h>
#include <salsa.h>
#include <poly1305.h>
#include <hex.h>
#include <osrng.h>
#include <filters.h>

using namespace CryptoPP;
using namespace std;

// Function to encrypt using Salsa20-Poly1305
string salsa20Poly1305Encrypt(const string& plaintext, const SecByteBlock& key, const byte nonce[8], const string& associatedData, string& tag) {
    string ciphertext;

    Salsa20::Encryption salsa;
    salsa.SetKeyWithIV(key, key.size(), nonce, 8); // Set key and nonce

    StringSource(plaintext, true,
        new StreamTransformationFilter(salsa,
            new StringSink(ciphertext)
        )
    );

    // Generate Poly1305 authentication tag
    Poly1305 poly;
    poly.SetKey(key, key.size());

    StringSource(associatedData + ciphertext, true,
        new HashFilter(poly,
            new StringSink(tag)
        )
    );

    return ciphertext;
}

// Function to decrypt using Salsa20-Poly1305
string salsa20Poly1305Decrypt(const string& encrypted, const SecByteBlock& key, const byte nonce[8], const string& associatedData, const string& tag) {
    string decrypted, computedTag;

    // Verify Poly1305 authentication tag
    Poly1305 poly;
    poly.SetKey(key, key.size());

    StringSource(associatedData + encrypted, true,
        new HashFilter(poly,
            new StringSink(computedTag)
        )
    );

    if (computedTag != tag) {
        throw runtime_error("ðŸ”´ Authentication failed!");
    }

    // Decrypt using Salsa20
    Salsa20::Decryption salsa;
    salsa.SetKeyWithIV(key, key.size(), nonce, 8); // Set key and nonce

    StringSource(encrypted, true,
        new StreamTransformationFilter(salsa,
            new StringSink(decrypted)
        )
    );

    return decrypted;
}

int main() {
    AutoSeededRandomPool prng;

    SecByteBlock key(Salsa20::DEFAULT_KEYLENGTH); // 256-bit key
    prng.GenerateBlock(key, key.size());

    byte nonce[8]; // 64-bit nonce
    prng.GenerateBlock(nonce, sizeof(nonce));

    string plaintext, associatedData = "AUTH_DATA", tag;
    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    // Encrypt
    string encrypted = salsa20Poly1305Encrypt(plaintext, key, nonce, associatedData, tag);
    cout << "ðŸ” Encrypted (Hex): ";
    StringSource(encrypted, true, new HexEncoder(new FileSink(cout)));
    cout << endl;

    // Decrypt
    try {
        string decrypted = salsa20Poly1305Decrypt(encrypted, key, nonce, associatedData, tag);
        cout << "ðŸ”“ Decrypted: " << decrypted << endl;
    } catch (const exception& e) {
        cout << e.what() << endl;
    }

    return 0;
}


# 17(b). **XSalsa20-Poly1305 Cipher**
## **Concept**
- **XSalsa20-Poly1305** is an **AEAD (Authenticated Encryption with Associated Data)** cipher.
- It extends **Salsa20** by using a **192-bit nonce (24 bytes)** instead of the standard 64-bit.
- **Poly1305** is used for message authentication to ensure data integrity.
- Widely used in **secure messaging protocols (Signal), VPNs, and modern cryptographic systems**.

---

## **Algorithm Steps**
1. Generate a **random 192-bit nonce (24 bytes)**.
2. Initialize **XSalsa20-Poly1305** with:
   - **256-bit key (32 bytes)**
   - **24-byte nonce**
   - **Optional associated data (AD)**
3. Encrypt plaintext using **XSalsa20** stream cipher.
4. Generate an **authentication tag** using **Poly1305**.
5. Decryption verifies the **authentication tag** before decrypting.
   - If verification fails, decryption is **rejected**.

---

## **C++ Implementation (libsodium)**
```cpp
#include <sodium.h>
#include <iostream>
#include <vector>

using namespace std;

// Function to encrypt using XSalsa20-Poly1305
vector<unsigned char> xsalsa20_poly1305_encrypt(const string& plaintext, const vector<unsigned char>& key, vector<unsigned char>& nonce) {
    nonce.resize(crypto_secretbox_NONCEBYTES);
    randombytes_buf(nonce.data(), crypto_secretbox_NONCEBYTES);

    vector<unsigned char> ciphertext(plaintext.size() + crypto_secretbox_MACBYTES);
    crypto_secretbox_easy(ciphertext.data(), (const unsigned char*)plaintext.data(), plaintext.size(), nonce.data(), key.data());

    return ciphertext;
}

// Function to decrypt using XSalsa20-Poly1305
string xsalsa20_poly1305_decrypt(const vector<unsigned char>& ciphertext, const vector<unsigned char>& key, const vector<unsigned char>& nonce) {
    vector<unsigned char> decrypted(ciphertext.size() - crypto_secretbox_MACBYTES);

    if (crypto_secretbox_open_easy(decrypted.data(), ciphertext.data(), ciphertext.size(), nonce.data(), key.data()) != 0) {
        throw runtime_error("ðŸ”´ Authentication failed!");
    }

    return string(decrypted.begin(), decrypted.end());
}

int main() {
    if (sodium_init() < 0) {
        cout << "Failed to initialize libsodium!" << endl;
        return 1;
    }

    vector<unsigned char> key(crypto_secretbox_KEYBYTES);
    randombytes_buf(key.data(), key.size());

    vector<unsigned char> nonce;
    string plaintext;
    
    cout << "Enter text to encrypt: ";
    getline(cin, plaintext);

    vector<unsigned char> encrypted = xsalsa20_poly1305_encrypt(plaintext, key, nonce);
    cout << "ðŸ” Encrypted (Hex): ";
    for (unsigned char c : encrypted) cout << hex << (int)c;
    cout << endl;

    try {
        string decrypted = xsalsa20_poly1305_decrypt(encrypted, key, nonce);
        cout << "ðŸ”“ Decrypted: " << decrypted << endl;
    } catch (const exception& e) {
        cout << e.what() << endl;
    }

    return 0;
}
