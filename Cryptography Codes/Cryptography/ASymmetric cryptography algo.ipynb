````python
from Crypto.PublicKey import RSA
from Crypto.Cipher import PKCS1_OAEP
import base64

# Function to generate RSA key pair
def generate_rsa_keys(key_size=2048):
    key = RSA.generate(key_size)  # Generate RSA key pair
    private_key = key.export_key()  # Export private key
    public_key = key.publickey().export_key()  # Export public key
    return private_key, public_key

# Function to encrypt using RSA public key
def rsa_encrypt(plaintext, public_key):
    recipient_key = RSA.import_key(public_key)  # Import public key
    cipher = PKCS1_OAEP.new(recipient_key)  # Initialize cipher
    encrypted_text = cipher.encrypt(plaintext.encode())  # Encrypt plaintext
    return base64.b64encode(encrypted_text).decode()  # Encode in Base64

# Function to decrypt using RSA private key
def rsa_decrypt(encrypted_text, private_key):
    private_key = RSA.import_key(private_key)  # Import private key
    cipher = PKCS1_OAEP.new(private_key)  # Initialize cipher
    decrypted_text = cipher.decrypt(base64.b64decode(encrypted_text))  # Decrypt
    return decrypted_text.decode()

# Example Usage
private_key, public_key = generate_rsa_keys()  # Generate key pair
plaintext = "HELLO123"

encrypted = rsa_encrypt(plaintext, public_key)
decrypted = rsa_decrypt(encrypted, private_key)

print("ğŸ” Encrypted (RSA):", encrypted)
print("ğŸ”“ Decrypted:", decrypted)
````

### **ğŸ”¹ How Does Diffie-Hellman Work?**  

Instead of **encrypting and decrypting messages**, Diffie-Hellman helps **two users establish a shared secret key** that can later be used for **symmetric encryption (AES, ChaCha20, etc.)**.  

ğŸ”¹ **Step-by-Step Process:**  

1ï¸âƒ£ **Public Parameters (Chosen Beforehand)**  
   - A **large prime number** `p`.

2ï¸âƒ£ **Each Party Generates a Private Key**  
   - **Alice** selects a private key `a` (random integer).  
   - **Bob** selects a private key `b` (random integer).  

3ï¸âƒ£ **Compute Public Keys (Sent Over an Insecure Channel)**  
   - Alice computes: **`A â‰¡ g^a mod p`** â†’ Sends **A** to Bob.  
   - Bob computes: **`B â‰¡ g^b mod p`** â†’ Sends **B** to Alice.  

4ï¸âƒ£ **Compute the Shared Secret (Without Exposing Private Keys)**  
   - Alice computes: **`S â‰¡ B^a mod p`**.

### **ğŸ”¹ How Does ElGamal Work?**  

ElGamal consists of **three main steps: Key Generation, Encryption, and Decryption.**  

ğŸ”¹ **1ï¸âƒ£ Key Generation**  
- Choose a **large prime number `p`** and a **generator `g`** (same as Diffie-Hellman).  
- Choose a **private key `x`** (random integer).

ğŸ”¹ **2ï¸âƒ£ Encryption (Using the Public Key)**  
- Choose a **random ephemeral key `k`**.

ğŸ”¹ **3ï¸âƒ£ Decryption (Using the Private Key)**  
- Compute **shared secret `s = c1^x mod p`**.

âœ… **Only the receiver (who knows `x`) can compute `s` and recover the plaintext `M`.**

âœ… **Provides semantic security â€“ different encryptions of the same message result in different ciphertexts.**

````python
from Crypto.Util.number import getPrime, inverse, GCD
import random

# Function to generate ElGamal keys
def generate_elgamal_keys(bit_length=2048):
    p = getPrime(bit_length)  # Generate a large prime number p
    g = random.randint(2, p - 2)  # Choose a generator g
    x = random.randint(2, p - 2)  # Private key (random integer)
    y = pow(g, x, p)  # Compute public key y = g^x mod p
    return (p, g, y), x  # Public key (p, g, y), Private key (x)

# Function to encrypt a message using ElGamal
def elgamal_encrypt(plaintext, public_key):
    p, g, y = public_key
    m = int.from_bytes(plaintext.encode(), 'big')  # Convert plaintext to an integer
    k = random.randint(2, p - 2)  # Random ephemeral key
    c1 = pow(g, k, p)  # Compute c1 = g^k mod p
    c2 = (m * pow(y, k, p)) % p  # Compute c2 = (M * y^k) mod p
    return c1, c2  # Ciphertext (c1, c2)

# Function to decrypt a message using ElGamal
def elgamal_decrypt(ciphertext, private_key, p):
    c1, c2 = ciphertext
    s = pow(c1, private_key, p)  # Compute shared secret s = c1^x mod p
    s_inv = inverse(s, p)  # Compute modular inverse of s
    m = (c2 * s_inv) % p  # Compute M = (c2 * s^(-1)) mod p
    return m.to_bytes((m.bit_length() + 7) // 8, 'big').decode()  # Convert integer back to string

# Generate ElGamal keys
public_key, private_key = generate_elgamal_keys()

# Example usage
plaintext = "HELLO123"
ciphertext = elgamal_encrypt(plaintext, public_key)
decrypted_text = elgamal_decrypt(ciphertext, private_key, public_key[0])

print("ğŸ” Encrypted (ElGamal):", ciphertext)
print("ğŸ”“ Decrypted:", decrypted_text)
````

# ğŸ“Œ **Project 4: Elliptic Curve Cryptography (ECC) â€“ A Deep Dive**  

## ğŸ“Œ **Introduction**  

Elliptic Curve Cryptography (ECC) is a modern asymmetric cryptographic system that provides the same level of security as traditional methods like RSA but with much smaller key sizes. ECC is widely used in secure communication protocols such as TLS, Signal, SSH, and cryptocurrency systems like Bitcoin.  

ECC is based on the mathematics of elliptic curves over finite fields and offers the following advantages:  
âœ… **Smaller Key Sizes** â†’ Strong security with fewer bits (e.g., 256-bit ECC is as secure as 3072-bit RSA).  
âœ… **Faster Computations** â†’ Operations are more efficient, leading to quicker encryption/decryption.  
âœ… **Lower Power Consumption** â†’ Ideal for resource-constrained devices like IoT and mobile.  

---  

## ğŸ“Œ **How ECC Works**  

ECC operates on the mathematical properties of elliptic curves defined by the equation:  

\[
y^2 = x^3 + ax + b
\]

where `a` and `b` are constants that define a specific elliptic curve. The security of ECC relies on the **Elliptic Curve Discrete Logarithm Problem (ECDLP)**, which is computationally hard to solve.  

### ğŸ”¹ **Key Operations in ECC**  

1ï¸âƒ£ **Key Generation** â†’ A private key `d` is chosen randomly, and the public key `Q` is computed as:  
   \[
   Q = dG
   \]
   where `G` is a predefined generator point on the curve.  

2ï¸âƒ£ **Encryption (Using ECIES - Elliptic Curve Integrated Encryption Scheme)**  
   - Choose a random ephemeral key `r`.  
   - Compute the shared point `P = r * Q`.  
   - Encrypt the plaintext using a symmetric cipher (e.g., AES) with a derived key from `P`.  

3ï¸âƒ£ **Decryption**  
   - Compute `P = d * R` (where `R` is received from the sender).  
   - Derive the symmetric key and decrypt the message.  

4ï¸âƒ£ **Key Exchange (ECDH - Elliptic Curve Diffie-Hellman)**  
   - Both parties exchange public keys and compute a shared secret:  
     \[
     S = d_A * Q_B = d_B * Q_A
     \]
   - The computed shared secret is used for symmetric encryption.  

5ï¸âƒ£ **Digital Signatures (ECDSA & EdDSA)**  
   - A private key signs a message, and the signature can be verified using the public key.  
   - Used for authentication and data integrity (e.g., blockchain, SSL/TLS).

# ğŸ“Œ **Project 5: ECDH (Elliptic Curve Diffie-Hellman) â€“ A Deep Dive**  

## ğŸ“Œ **Introduction**  

Elliptic Curve Diffie-Hellman (**ECDH**) is a **key exchange protocol** that allows two parties to securely derive a **shared secret** over an insecure channel. It is an elliptic curve variant of the classic **Diffie-Hellman (DH) key exchange** but provides the same security with much **smaller key sizes**.  

ECDH is widely used in:  
âœ… **TLS/SSL Handshakes** â†’ Establishes secure connections in HTTPS.  
âœ… **Encrypted Messaging** â†’ Used in Signal, WhatsApp, and other end-to-end encrypted apps.  
âœ… **Cryptocurrency Wallets** â†’ Protects private keys in blockchain transactions.  

### **Why Use ECDH Instead of Classical DH?**  
ğŸ”¹ **Stronger Security with Smaller Keys** â†’ A 256-bit ECDH key provides the same security as a 3072-bit DH key.  
ğŸ”¹ **Faster Computation** â†’ ECC operations are more efficient than modular exponentiation in DH.  
ğŸ”¹ **Ideal for Constrained Devices** â†’ Used in IoT and embedded systems due to its efficiency.  

---

## ğŸ“Œ **How ECDH Works**  

ECDH enables two parties, **Alice** and **Bob**, to derive a **shared secret key** without directly transmitting it. The security of ECDH relies on the **Elliptic Curve Discrete Logarithm Problem (ECDLP)**, making it infeasible for attackers to compute the private keys from public information.  

### ğŸ”¹ **Steps in ECDH Key Exchange**  

1ï¸âƒ£ **Key Pair Generation (Both Alice & Bob)**  
   - Each party generates an **ECC private key** (`dA` for Alice, `dB` for Bob).  
   - Computes their respective **public keys**:  
     \[
     P_A = d_A G, \quad P_B = d_B G
     \]
     where `G` is a fixed generator point on the elliptic curve.  

2ï¸âƒ£ **Public Key Exchange**  
   - Alice and Bob exchange their public keys (`P_A` and `P_B`).  

3ï¸âƒ£ **Shared Secret Computation**  
   - Each party computes the **same shared secret**:  
     \[
     S = d_A P_B = d_B P_A
     \]
   - Since `P_B = d_B G`, Aliceâ€™s computation simplifies to:  
     \[
     S = d_A (d_B G) = (d_A d_B) G
     \]
   - Similarly, Bob computes:  
     \[
     S = d_B (d_A G) = (d_A d_B) G
     \]
   - Both derive the same **shared secret key** without directly transmitting private keys!  

ğŸš€ **Key Idea:** Even if an attacker intercepts the public keys `P_A` and `P_B`, they **cannot compute `S`** without knowing `d_A` or `d_B`, thanks to the difficulty of solving **ECDLP**.

# ğŸ“Œ **Project 6: ECIES (Elliptic Curve Integrated Encryption Scheme) â€“ A Deep Dive**  

## ğŸ“Œ **Introduction**  

Elliptic Curve Integrated Encryption Scheme (**ECIES**) is a **hybrid encryption scheme** that combines **Elliptic Curve Cryptography (ECC)** with **symmetric encryption** and **message authentication** to securely encrypt data.  

ğŸš€ **Why ECIES?**  
ECIES is designed to provide **both confidentiality and integrity** by using:  
âœ… **ECC for key exchange** (small key sizes, high security).  
âœ… **Symmetric encryption (AES, ChaCha20)** for fast data encryption.  
âœ… **Key Derivation Function (KDF)** to generate a secure encryption key.  
âœ… **Message Authentication Code (MAC)** to ensure message integrity.  

### **Real-World Uses of ECIES**  
ğŸ”¹ **Secure Encrypted Messaging** â†’ Used in PGP and end-to-end encrypted applications.  
ğŸ”¹ **Cryptocurrency Wallets** â†’ Protects private keys in Ethereum and Bitcoin.  
ğŸ”¹ **TLS Handshakes** â†’ Used in securing web communication.  

---

## ğŸ“Œ **How ECIES Works**  

ECIES follows a **hybrid encryption approach**, where:  
- **ECC is used to establish a shared secret key.**  
- **A symmetric cipher (AES/ChaCha20) encrypts the actual message.**  
- **A MAC ensures the integrity of the encrypted message.**  

### ğŸ”¹ **Step-by-Step Working of ECIES**  

1ï¸âƒ£ **Key Pair Generation (Sender & Receiver)**  
   - Receiver generates an **ECC private key** (`dR`) and **public key** (`P_R = dR * G`).  
   - Sender generates an ephemeral ECC **private key** (`dE`) and computes the corresponding **public key** (`P_E = dE * G`).  

2ï¸âƒ£ **Key Exchange Using ECC**  
   - Sender computes a **shared secret** using the receiverâ€™s public key:  
     \[
     S = dE * P_R
     \]  
   - This ensures that only the receiver (who knows `dR`) can compute the **same shared secret**.  

3ï¸âƒ£ **Key Derivation & Encryption**  
   - A **Key Derivation Function (KDF)** (e.g., HKDF, SHA-256) expands `S` into:  
     - **Symmetric encryption key** (for AES/ChaCha20).  
     - **MAC key** (for message authentication).  
   - The message is **encrypted** using AES-GCM or ChaCha20.  
   - A **MAC is computed** over the ciphertext to ensure integrity.  

4ï¸âƒ£ **Decryption by the Receiver**  
   - Receiver computes the **shared secret** using:  
     \[
     S = dR * P_E
     \]  
   - Uses the **KDF** to derive the encryption & MAC keys.  
   - **Verifies the MAC** and decrypts the message.  

ğŸš€ **Key Idea:** Even if an attacker intercepts the ephemeral public key (`P_E`), they **cannot compute `S`** without the receiverâ€™s private key `dR`.

# ğŸ”¹ 1. Generate Key Pairs for Sender & Receiver

````python
from cryptography.hazmat.primitives.asymmetric import ec

# Generate receiver's key pair
receiver_private_key = ec.generate_private_key(ec.SECP256R1())
receiver_public_key = receiver_private_key.public_key()

# Generate sender's ephemeral key pair
sender_private_key = ec.generate_private_key(ec.SECP256R1())
sender_public_key = sender_private_key.public_key()

print("Receiver's Public Key:", receiver_public_key)
print("Sender's Ephemeral Public Key:", sender_public_key)
````

# ğŸ”¹ 2. Compute the Shared Secret & Derive Keys

````python
from cryptography.hazmat.primitives.kdf.hkdf import HKDF
from cryptography.hazmat.primitives.hashes import SHA256

# Sender computes shared secret using receiver's public key
shared_secret = sender_private_key.exchange(ec.ECDH(), receiver_public_key)

# Derive AES and MAC keys using HKDF
derived_key = HKDF(
    algorithm=SHA256(),
    length=32,  # 16 bytes for AES, 16 for MAC
    salt=None,
    info=b"ecies key agreement"
).derive(shared_secret)

aes_key = derived_key[:16]  # First 16 bytes for AES
mac_key = derived_key[16:]  # Last 16 bytes for MAC

print("Derived AES Key:", aes_key.hex())
print("Derived MAC Key:", mac_key.hex())
````

# ğŸ”¹ 3. Encrypt the Message Using AES-GCM

````python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
import os

# Message to encrypt
plaintext = b"Hello, this is a secret message!"

# Generate a random IV (nonce)
iv = os.urandom(12)

# Encrypt using AES-GCM
cipher = Cipher(algorithms.AES(aes_key), modes.GCM(iv))
encryptor = cipher.encryptor()
ciphertext = encryptor.update(plaintext) + encryptor.finalize()

# Get authentication tag
tag = encryptor.tag

print("Ciphertext:", ciphertext.hex())
print("Authentication Tag:", tag.hex())
````

# ğŸ”¹ 4. Decrypt the Message & Verify Integrity

````python
# Receiver reconstructs AES key using same process
receiver_shared_secret = receiver_private_key.exchange(ec.ECDH(), sender_public_key)
receiver_derived_key = HKDF(
    algorithm=SHA256(),
    length=32,
    salt=None,
    info=b"ecies key agreement"
).derive(receiver_shared_secret)

receiver_aes_key = receiver_derived_key[:16]  # AES Key
receiver_mac_key = receiver_derived_key[16:]  # MAC Key

# Decrypt the message
decrypt_cipher = Cipher(algorithms.AES(receiver_aes_key), modes.GCM(iv, tag))
decryptor = decrypt_cipher.decryptor()
decrypted_message = decryptor.update(ciphertext) + decryptor.finalize()

print("Decrypted Message:", decrypted_message.decode())
````
