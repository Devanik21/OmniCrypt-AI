## ğŸ”¹ 1. Brute-Force Attack (Trying All Possible Shifts)
- Since the Caesar cipher has only **25 possible shifts** (excluding shift=0), a hacker can **try all shifts** until they find readable text.
- This method is **guaranteed to work** and is extremely fast.

ğŸ›  **Hacker Mindset:**
- Visually inspect all 25 outputs and find the one that makes sense.

â³ **Time Complexity:** **O(1)** (Fixed number of possibilities).

 def caesar_cipher(text, shift):
     result = ""
     for char in text:
         if char.isalpha():
             shift_base = ord('A') if char.isupper() else ord('a')

## ğŸ”¹ 3. Known Plaintext Attack (KPA)
- If a hacker knows or suspects part of the original message (e.g., the text starts with "HELLO"), they can **reverse-engineer the shift**.
- Example:
  - If â€˜Hâ€™ in plaintext became â€˜Mâ€™ in ciphertext â†’ The shift is **+5**.
  - Apply this shift in reverse to decrypt the entire message.

ğŸ” **Why This Works?**
- Many encrypted messages follow predictable formats (e.g., greetings, signatures, common phrases).
- If even **one word is guessed**, the rest of the text can be cracked instantly.

ğŸ›  **Hacker Mindset:**
- Find a known or repeated word â†’ Compare it to the ciphertext â†’ Determine shift â†’ Decrypt.

â³ **Time Complexity:** **O(1)** (If known plaintext is present).
