# ğŸš€ One-Time Pad (OTP) Cipher â€“ The Unbreakable Cipher

The One-Time Pad (OTP) is the only provably unbreakable cipher, but it requires a truly random key that is the same length as the message.


ğŸ”¹ How One-Time Pad Works

1ï¸âƒ£ Generate a random key (same 
length as the plaintext).

2ï¸âƒ£ Encrypt: Perform a bitwise 
XOR between the plaintext and the 
key.

3ï¸âƒ£ Decrypt: Perform XOR again 
with the same key to retrieve the

âœ… Perfect secrecy (if the key is truly random & never reused).

âŒ Impractical for large-scale use (key must be as long as the

```python
import os

# Function to convert text to binary
def text_to_binary(text):
    return ''.join(format(ord(char), '08b') for char in text)

# Function to convert binary to text
def binary_to_text(binary):
    return ''.join(chr(int(binary[i:i+8], 2)) for i in range(0, len(binary), 8))

# Function to generate a truly random key of the same length as the message
def generate_key(length):
    return os.urandom(length)  # Generates cryptographically secure random bytes

# Function to encrypt using One-Time Pad (XOR operation)
def otp_encrypt(plaintext):
    plaintext_bytes = plaintext.encode()
    key = generate_key(len(plaintext_bytes))
    ciphertext = bytes([p ^ k for p, k in zip(plaintext_bytes, key)])
    return key, ciphertext

# Function to decrypt using One-Time Pad
def otp_decrypt(ciphertext, key):
    decrypted_bytes = bytes([c ^ k for c, k in zip(ciphertext, key)])
    return decrypted_bytes.decode()

# Get user input
plaintext = input("Enter text to encrypt: ")

# Encrypt
key, encrypted_text = otp_encrypt(plaintext)
print("ğŸ” Encrypted (Hex):", encrypted_text.hex())
print("ğŸ— Key (Hex):", key.hex())  # Key must be securely shared

# Decrypt
decrypted_text = otp_decrypt(encrypted_text, key)
print("ğŸ”“ Decrypted:", decrypted_text)
```

ğŸ”¹ What is a Rotor Cipher?
