Since you're starting with cryptography, a simple project is implementing Caesar Cipherâ€”one of the oldest encryption techniques. Here's how it works:

Each letter in the plaintext is shifted by a fixed number (key).


For example, if the key is 3, then A â†’ D, B â†’ E, C â†’ F, and so on.

Decryption is just shifting in the opposite direction.

# ğŸ”¹ Project 1 : Caesar Cipher in Python

````python
def caesar_cipher(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            shift_base = ord('A') if char.isupper() else ord('a')
            result += chr((ord(char) - shift_base + shift) % 26 + shift_base)
        else:
            result += char  # Keep non-alphabet characters unchanged
    return result

# Encrypt
text = input("Enter text: ")
shift = int(input("Enter shift key: "))
encrypted_text = caesar_cipher(text, shift)
print("Encrypted:", encrypted_text)

# Decrypt
decrypted_text = caesar_cipher(encrypted_text, -shift)
print("Decrypted:", decrypted_text)
````

# ğŸ”¹ Project 4 : VigenÃ¨re Cipher

````python
def vigenere_cipher(text, key, encrypt=True):
    result = []
    key = key.lower()
    key_length = len(key)

    for i, char in enumerate(text):
        if char.isalpha():
            shift = ord(key[i % key_length]) - ord('a')  # Get shift from keyword
            if not encrypt:
                shift = -shift  # Reverse shift for decryption

            shift_base = ord('A') if char.isupper() else ord('a')
            result.append(chr((ord(char) - shift_base + shift) % 26 + shift_base))
        else:
            result.append(char)  # Keep non-alphabet characters unchanged

    return "".join(result)

# Get user input
text = input("Enter text: ")
key = input("Enter keyword: ")

# Encrypt
encrypted_text = vigenere_cipher(text, key)
print("Encrypted:", encrypted_text)

# Decrypt
decrypted_text = vigenere_cipher(encrypted_text, key, encrypt=False)
print("Decrypted:", decrypted_text)
````

# ğŸ”¹ Project 5: Playfair Cipher in Python

````python
import numpy as np

def generate_playfair_matrix(key):
    key = "".join(dict.fromkeys(key + "ABCDEFGHIKLMNOPQRSTUVWXYZ"))  # Remove duplicates & add missing letters
    matrix = np.array(list(key)).reshape(5, 5)
    return matrix

def find_position(matrix, letter):
    pos = np.where(matrix == letter)
    return pos[0][0], pos[1][0]

def playfair_cipher(text, matrix, encrypt=True):
    text = text.upper().replace("J", "I")  # Replace J with I
    text = [char for char in text if char.isalpha()]  # Remove non-alphabet chars
    
    # Create digraphs
    digraphs = []
    i = 0
    while i < len(text):
        if i + 1 < len(text) and text[i] != text[i + 1]:
            digraphs.append(text[i] + text[i + 1])
            i += 2
        else:
            digraphs.append(text[i] + "X")
            i += 1

    result = []
    shift = 1 if encrypt else -1

    for a, b in digraphs:
        r1, c1 = find_position(matrix, a)
        r2, c2 = find_position(matrix, b)
        
        if r1 == r2:  # Same row
            result.append(matrix[r1, (c1 + shift) % 5] + matrix[r2, (c2 + shift) % 5])
        elif c1 == c2:  # Same column
            result.append(matrix[(r1 + shift) % 5, c1] + matrix[(r2 + shift) % 5, c2])
        else:  # Rectangle swap
            result.append(matrix[r1, c2] + matrix[r2, c1])

    return "".join(result)

# Secret Key
key = "CRYPTO"
matrix = generate_playfair_matrix(key)

# Get user input
text = input("Enter text: ")

# Encrypt
encrypted_text = playfair_cipher(text, matrix)
print("Encrypted:", encrypted_text)

# Decrypt
decrypted_text = playfair_cipher(encrypted_text, matrix, encrypt=False)
print("Decrypted:", decrypted_text)


````

# ğŸ”¹ Project 6: Rail Fence Cipher (Zigzag Transposition Cipher)

ğŸ”¹ What is Rail Fence Cipher?

It is a transposition cipher (rearranges letters instead of substituting them).

Encrypts by writing the text in a zigzag pattern across multiple "rails" (rows).

Key = Number of Rails (rows used in encryption).

Decryption reconstructs the zigzag pattern to get the original text.

ğŸ”¹ Step 1: Secret Key for Rail Fence Cipher



# ğŸ”‘ Secret Key for Rail Fence Cipher
**Number of Rails (Key):** `3`

````python
def rail_fence_encrypt(text, key):
    if key <= 1:
        return text  # No encryption needed

    # Create empty rail matrix
    rail = [['\n' for _ in range(len(text))] for _ in range(key)]

    row, step = 0, 1
    for i in range(len(text)):
        rail[row][i] = text[i]
        if row == 0:
            step = 1
        elif row == key - 1:
            step = -1
        row += step

    # Read the matrix row-wise
    return ''.join(char for row in rail for char in row if char != '\n')


def rail_fence_decrypt(ciphertext, key):
    if key <= 1:
        return ciphertext  # No decryption needed

    # Create empty rail matrix
    rail = [['\n' for _ in range(len(ciphertext))] for _ in range(key)]

    # Mark zigzag positions
    row, step = 0, 1
    for i in range(len(ciphertext)):
        rail[row][i] = '*'
        if row == 0:
            step = 1
        elif row == key - 1:
            step = -1
        row += step

    # Fill the rail matrix with ciphertext characters
    index = 0
    for r in range(key):
        for c in range(len(ciphertext)):
            if rail[r][c] == '*' and index < len(ciphertext):
                rail[r][c] = ciphertext[index]
                index += 1

    # Read the matrix in zigzag order
    row, step, decrypted_text = 0, 1, []
    for i in range(len(ciphertext)):
        decrypted_text.append(rail[row][i])
        if row == 0:
            step = 1
        elif row == key - 1:
            step = -1
        row += step

    return ''.join(decrypted_text)


# Get user input
text = input("Enter text to encrypt: ")
rails = int(input("Enter number of rails (key): "))

# Encrypt
encrypted_text = rail_fence_encrypt(text, rails)
print("ğŸ” Encrypted:", encrypted_text)

# Decrypt
decrypted_text = rail_fence_decrypt(encrypted_text, rails)
print("ğŸ”“ Decrypted:", decrypted_text)
````

````python
import math

def columnar_encrypt(text, key):
    key_order = sorted(list(enumerate(key)), key=lambda x: x[1])
    num_cols = len(key)
    num_rows = math.ceil(len(text) / num_cols)
    
    # Fill grid row-wise
    grid = [['' for _ in range(num_cols)] for _ in range(num_rows)]
    index = 0
    for r in range(num_rows):
        for c in range(num_cols):
            if index < len(text):
                grid[r][c] = text[index]
                index += 1

    # Read column-wise in key order
    ciphertext = ''.join(grid[r][c] for c, _ in key_order for r in range(num_rows) if grid[r][c])
    return ciphertext

def columnar_decrypt(ciphertext, key):
    key_order = sorted(list(enumerate(key)), key=lambda x: x[1])
    num_cols = len(key)
    num_rows = math.ceil(len(ciphertext) / num_cols)

    # Fill empty grid
    grid = [['' for _ in range(num_cols)] for _ in range(num_rows)]
    index = 0
    for c, _ in key_order:
        for r in range(num_rows):
            if index < len(ciphertext):
                grid[r][c] = ciphertext[index]
                index += 1

    # Read row-wise
    plaintext = ''.join(''.join(row) for row in grid)
    return plaintext

# Get user input
text = input("Enter text to encrypt: ")
key = input("Enter numerical key (e.g., 34152): ")

# Encrypt
encrypted_text = columnar_encrypt(text, key)
print("ğŸ” Encrypted:", encrypted_text)

# Decrypt
decrypted_text = columnar_decrypt(encrypted_text, key)
print("ğŸ”“ Decrypted:", decrypted_text)


````

# ğŸš€ Project 8: Hill Cipher (Matrix-Based Encryption)
Now, we move to a math-heavy but powerful cipher: Hill Cipher! ğŸ”¥



ğŸ”¹ What is Hill Cipher?

Uses matrix multiplication for encryption.

Unlike previous ciphers, it encrypts multiple letters at a time using linear algebra.

Key is a square matrix (e.g., 2Ã—2 or 3Ã—3) that must be invertible for decryption.

More secure than simple transposition/substitution ciphers.

ğŸ”¹ This matrix will be used for mod 26 matrix multiplication to encrypt letters.



````python
import numpy as np

# Convert letter to number (A=0, B=1, ..., Z=25)
def text_to_numbers(text):
    return [ord(char) - ord('A') for char in text]

# Convert number to letter
def numbers_to_text(numbers):
    return ''.join(chr(num + ord('A')) for num in numbers)

# Encrypt using Hill Cipher
def hill_encrypt(plaintext, key_matrix):
    plaintext = plaintext.upper().replace(" ", "")  # Remove spaces & uppercase
    if len(plaintext) % len(key_matrix) != 0:
        plaintext += "X"  # Padding if not a multiple of matrix size
    
    numbers = text_to_numbers(plaintext)
    encrypted_numbers = []
    
    # Encrypt in blocks
    for i in range(0, len(numbers), len(key_matrix)):
        block = np.array(numbers[i:i+len(key_matrix)]).reshape(-1, 1)
        encrypted_block = np.dot(key_matrix, block) % 26
        encrypted_numbers.extend(encrypted_block.flatten())
    
    return numbers_to_text(encrypted_numbers)

# Find modular inverse of key matrix
def modular_inverse(matrix, mod=26):
    det = int(np.round(np.linalg.det(matrix)))  # Determinant
    det_inv = pow(det, -1, mod)  # Modular inverse of determinant
    
    adjugate = np.round(det * np.linalg.inv(matrix)).astype(int) % mod  # Adjugate matrix
    return (det_inv * adjugate) % mod  # Inverse matrix mod 26

# Decrypt using Hill Cipher
def hill_decrypt(ciphertext, key_matrix):
    key_inverse = modular_inverse(key_matrix)
    numbers = text_to_numbers(ciphertext)
    decrypted_numbers = []
    
    # Decrypt in blocks
    for i in range(0, len(numbers), len(key_matrix)):
        block = np.array(numbers[i:i+len(key_matrix)]).reshape(-1, 1)
        decrypted_block = np.dot(key_inverse, block) % 26
        decrypted_numbers.extend(decrypted_block.flatten())
    
    return numbers_to_text(decrypted_numbers)

# 2Ã—2 Key Matrix
key_matrix = np.array([[3, 3], [2, 5]])

# Get user input
plaintext = input("Enter text to encrypt: ")

# Encrypt
encrypted_text = hill_encrypt(plaintext, key_matrix)
print("ğŸ” Encrypted:", encrypted_text)

# Decrypt
decrypted_text = hill_decrypt(encrypted_text, key_matrix)
print("ğŸ”“ Decrypted:", decrypted_text)
````

ğŸ”¹ What is DES?

First widely adopted block cipher (developed by IBM & standardized in 1977).

Works with a 56-bit secret key (effectively 64 bits, but 8 bits are parity).

````python
from Crypto.Cipher import DES
import base64

# Function to ensure the key is exactly 8 bytes long
def format_key(key):
    if len(key) < 8:
        key = key.ljust(8, 'X')  # Pad with 'X' if too short
    elif len(key) > 8:
        key = key[:8]  # Trim if too long
    return key.encode()  # Convert to bytes

# Function to pad the text to be a multiple of 8 bytes
def pad(text):
    while len(text) % 8 != 0:
        text += " "  # Pad with spaces
    return text

# Function to encrypt using DES
def des_encrypt(plaintext, key):
    cipher = DES.new(key, DES.MODE_ECB)  # Using ECB mode
    padded_text = pad(plaintext).encode()
    encrypted_bytes = cipher.encrypt(padded_text)
    return base64.b64encode(encrypted_bytes).decode()

# Function to decrypt using DES
def des_decrypt(ciphertext, key):
    cipher = DES.new(key, DES.MODE_ECB)
    decrypted_bytes = cipher.decrypt(base64.b64decode(ciphertext))
    return decrypted_bytes.decode().strip()  # Remove padding

# Get user input
plaintext = input("Enter text to encrypt: ")
user_key = input("Enter an 8-character key: ")

# Format the key to ensure it's exactly 8 bytes
key = format_key(user_key)

# Encrypt
encrypted_text = des_encrypt(plaintext, key)
print("ğŸ” Encrypted:", encrypted_text)

# Decrypt
decrypted_text = des_decrypt(encrypted_text, key)
print("ğŸ”“ Decrypted:", decrypted_text)
````

# ğŸš€ Project 10: Triple DES (3DES) â€“ Stronger Version of DES

Now, we improve DES by implementing Triple DES (3DES), which applies DES encryption

ğŸ”¹ What is Triple DES (3DES)?

Instead of using one DES encryption, we apply it three times:


Encryption â†’ Decryption â†’ Encryption (EDE Mode).

Uses two or three 8-byte keys (total 16 or 24 bytes).

Much stronger than DES, but slower than AES.

````python
from Crypto.Cipher import DES3
import base64

# Function to ensure the key is 16 or 24 bytes
def format_key(key):
    if len(key) < 16:
        key = key.ljust(16, 'X')  # Pad with 'X' if too short
    elif len(key) > 24:
        key = key[:24]  # Trim if too long
    return key.encode()  # Convert to bytes

# Function to pad the text to be a multiple of 8 bytes
def pad(text):
    while len(text) % 8 != 0:
        text += " "  # Pad with spaces
    return text

# Function to encrypt using 3DES
def triple_des_encrypt(plaintext, key):
    cipher = DES3.new(key, DES3.MODE_ECB)  # Using ECB mode
    padded_text = pad(plaintext).encode()
    encrypted_bytes = cipher.encrypt(padded_text)
    return base64.b64encode(encrypted_bytes).decode()

# Function to decrypt using 3DES
def triple_des_decrypt(ciphertext, key):
    cipher = DES3.new(key, DES3.MODE_ECB)
    decrypted_bytes = cipher.decrypt(base64.b64decode(ciphertext))
    return decrypted_bytes.decode().strip()  # Remove padding

# Get user input
plaintext = input("Enter text to encrypt: ")
user_key = input("Enter a 16 or 24-character key: ")

# Format the key to ensure it's exactly 16 or 24 bytes
key = format_key(user_key)

# Encrypt
encrypted_text = triple_des_encrypt(plaintext, key)
print("ğŸ” Encrypted:", encrypted_text)

# Decrypt
decrypted_text = triple_des_decrypt(encrypted_text, key)
print("ğŸ”“ Decrypted:", decrypted_text)
````

# ğŸš€ Project 11: AES (Advanced Encryption Standard) â€“ The Most Secure Block Cipher

What is AES?

A modern symmetric encryption algorithm used in banking, Wi-Fi security (WPA2), 

HTTPS, and more.

Works on fixed-size blocks (128 bits = 16 bytes).

Uses 128-bit, 192-bit, or 256-bit keys.

Fast & highly secure compared to older ciphers like DES.

````python
from Crypto.Cipher import AES
import base64
import os

# Function to pad the text to make it 16-byte (128-bit) aligned
def pad(text):
    return text + (16 - len(text) % 16) * chr(16 - len(text) % 16)

# Function to remove padding after decryption
def unpad(text):
    return text[:-ord(text[-1])]

# Generate a random 16-byte (128-bit) key
key = os.urandom(16)  # Secret key (must be shared between sender & receiver)

def encrypt(plaintext, key):
    cipher = AES.new(key, AES.MODE_ECB)  # Using AES in ECB mode
    encrypted_text = cipher.encrypt(pad(plaintext).encode())
    return base64.b64encode(encrypted_text).decode()  # Encode output in Base64

def decrypt(encrypted_text, key):
    cipher = AES.new(key, AES.MODE_ECB)
    decrypted_text = cipher.decrypt(base64.b64decode(encrypted_text)).decode()
    return unpad(decrypted_text)

# Get user input
plaintext = input("Enter text to encrypt: ")

# Encrypt
encrypted_text = encrypt(plaintext, key)
print("Encrypted:", encrypted_text)

# Decrypt
decrypted_text = decrypt(encrypted_text, key)
print("Decrypted:", decrypted_text)
````

````python
import struct
from Crypto.Util.Padding import pad, unpad

# --- Helper functions for bit rotations ---
def rol(x, n, w=32):
    """Rotate left x by n bits for a word of width w bits."""
    return ((x << n) & ((1 << w) - 1)) | (x >> (w - n))

def ror(x, n, w=32):
    """Rotate right x by n bits for a word of width w bits."""
    return ((x >> n) | (x << (w - n))) & ((1 << w) - 1)

# --- RC6 Key Schedule ---
def rc6_key_schedule(key_bytes):
    """
    Generates the subkey array S from the given key bytes.
    Parameters:
      key_bytes (bytes): The encryption key.
    Returns:
      List of 32-bit subkeys S.
    """
    w = 32           # word size in bits
    r = 20           # number of rounds
    b = len(key_bytes)
    c = (b + 3) // 4 # number of 32-bit words in the key
    # Convert key to c little-endian 32-bit words (pad if needed)
    L = list(struct.unpack('<' + 'I' * c, key_bytes.ljust(c * 4, b'\0')))
    t = 2 * r + 4    # total number of subkeys = 44
    P32 = 0xB7E15163
    Q32 = 0x9E3779B9
    S = [0] * t
    S[0] = P32
    for i in range(1, t):
        S[i] = (S[i - 1] + Q32) & 0xffffffff

    i = j = 0
    A = B = 0
    # Mix the secret key into the S array
    for _ in range(3 * max(t, c)):
        A = S[i] = rol((S[i] + A + B) & 0xffffffff, 3)
        B = L[j] = rol((L[j] + A + B) & 0xffffffff, (A + B) % 32)
        i = (i + 1) % t
        j = (j + 1) % c
    return S

# --- RC6 Block Encryption (16-byte blocks) ---
def rc6_encrypt_block(block, S):
    """
    Encrypts a single 16-byte block using RC6.
    Parameters:
      block (bytes): 16-byte plaintext block.
      S (list): The subkeys generated by the key schedule.
    Returns:
      bytes: 16-byte ciphertext block.
    """
    w = 32
    r = 20
    # Unpack block into four 32-bit little-endian words: A, B, C, D
    A, B, C, D = struct.unpack('<IIII', block)
    B = (B + S[1]) & 0xffffffff
    D = (D + S[2]) & 0xffffffff
    for i in range(1, r + 1):
        t = rol((B * ((2 * B + 1) & 0xffffffff)) & 0xffffffff, 5)
        u = rol((D * ((2 * D + 1) & 0xffffffff)) & 0xffffffff, 5)
        A = (rol((A ^ t) & 0xffffffff, u & 0x1f) + S[2 * i + 2]) & 0xffffffff
        C = (rol((C ^ u) & 0xffffffff, t & 0x1f) + S[2 * i + 3]) & 0xffffffff
        A, B, C, D = B, C, D, A  # Rotate registers
    A = (A + S[2 * r + 2]) & 0xffffffff
    C = (C + S[2 * r + 3]) & 0xffffffff
    return struct.pack('<IIII', A, B, C, D)

# --- RC6 Block Decryption ---
def rc6_decrypt_block(block, S):
    """
    Decrypts a single 16-byte block using RC6.
    Parameters:
      block (bytes): 16-byte ciphertext block.
      S (list): The subkeys generated by the key schedule.
    Returns:
      bytes: 16-byte plaintext block.
    """
    w = 32
    r = 20
    A, B, C, D = struct.unpack('<IIII', block)
    C = (C - S[2 * r + 3]) & 0xffffffff
    A = (A - S[2 * r + 2]) & 0xffffffff
    for i in range(r, 0, -1):
        # Inverse rotate the registers
        A, B, C, D = D, A, B, C
        u = rol((D * ((2 * D + 1) & 0xffffffff)) & 0xffffffff, 5)
        t = rol((B * ((2 * B + 1) & 0xffffffff)) & 0xffffffff, 5)
        C = ror((C - S[2 * i + 3]) & 0xffffffff, t & 0x1f) ^ u
        A = ror((A - S[2 * i + 2]) & 0xffffffff, u & 0x1f) ^ t
    D = (D - S[2]) & 0xffffffff
    B = (B - S[1]) & 0xffffffff
    return struct.pack('<IIII', A, B, C, D)

# --- RC6 High-Level Encryption & Decryption ---
def rc6_encrypt(plaintext, key):
    """
    Encrypts an arbitrary-length plaintext using RC6.
    Parameters:
      plaintext (str): The plaintext string.
      key (str): The encryption key.
    Returns:
      str: Hexadecimal string of the ciphertext.
    """
    plaintext_bytes = plaintext.encode()
    # Pad plaintext to a multiple of 16 bytes (RC6 block size)
    plaintext_bytes = pad(plaintext_bytes, 16)
    key_bytes = key.encode()
    S = rc6_key_schedule(key_bytes)
    ciphertext = b''
    # Encrypt each 16-byte block
    for i in range(0, len(plaintext_bytes), 16):
        block = plaintext_bytes[i:i+16]
        ciphertext += rc6_encrypt_block(block, S)
    return ciphertext.hex()

def rc6_decrypt(ciphertext_hex, key):
    """
    Decrypts the RC6 ciphertext.
    Parameters:
      ciphertext_hex (str): Hexadecimal string of the ciphertext.
      key (str): The encryption key.
    Returns:
      str: The decrypted plaintext.
    """
    ciphertext = bytes.fromhex(ciphertext_hex)
    key_bytes = key.encode()
    S = rc6_key_schedule(key_bytes)
    decrypted = b''
    # Decrypt each 16-byte block
    for i in range(0, len(ciphertext), 16):
        block = ciphertext[i:i+16]
        decrypted += rc6_decrypt_block(block, S)
    decrypted = unpad(decrypted, 16)
    return decrypted.decode()

# --- Example Usage ---
if __name__ == '__main__':
    key = "MYRC6KEY1234567"  # Example key (typically 16 bytes works well)
    plaintext = "HELLO123"
    
    encrypted = rc6_encrypt(plaintext, key)
    decrypted = rc6_decrypt(encrypted, key)
    
    print("ğŸ” Encrypted (RC6):", encrypted)
    print("ğŸ”“ Decrypted:", decrypted)
````

# ğŸš€ Project 15: RC4 â€“ The Classic Stream Cipher ğŸ”¥ğŸ’¨  

RC4 (**Rivest Cipher 4**) is one of the most famous **stream ciphers** ever designed. It was widely used in **WEP, TLS, and VPN encryption**, but due to security vulnerabilities, it is now considered **weak** and is being phased out.

## **ğŸ”¹ What is RC4?**

âœ… **Extremely fast and efficient for software-based encryption.**

âŒ **Deprecated in TLS, Wi-Fi security (WEP), and modern cryptography.**  

---

````python
from Crypto.Cipher import ChaCha20
import base64
import os

# Function to encrypt using ChaCha20
def chacha20_encrypt(plaintext, key):
    key = key[:32]  # Ensure the key is 256-bit (32 bytes)
    cipher = ChaCha20.new(key=key)  # Initialize ChaCha20 cipher
    ciphertext = cipher.encrypt(plaintext.encode())  # Encrypt plaintext
    nonce = cipher.nonce  # Get nonce used in encryption
    return base64.b64encode(nonce + ciphertext).decode()  # Store nonce + ciphertext

# Function to decrypt using ChaCha20
def chacha20_decrypt(encrypted_text, key):
    key = key[:32]  # Ensure the key is 256-bit (32 bytes)
    encrypted_data = base64.b64decode(encrypted_text)
    nonce = encrypted_data[:8]  # Extract nonce (first 8 bytes)
    ciphertext = encrypted_data[8:]  # Extract ciphertext
    cipher = ChaCha20.new(key=key, nonce=nonce)  # Reinitialize cipher with the same nonce
    decrypted_text = cipher.decrypt(ciphertext)  # Decrypt the text
    return decrypted_text.decode()

# Example Usage
key = os.urandom(32)  # Generate a random 256-bit (32-byte) key
plaintext = "HELLO123"

encrypted = chacha20_encrypt(plaintext, key)
decrypted = chacha20_decrypt(encrypted, key)

print("ğŸ” Encrypted (ChaCha20):", encrypted)
print("ğŸ”“ Decrypted:", decrypted)
````

````python
from Crypto.Cipher import ChaCha20_Poly1305
from Crypto.Random import get_random_bytes
import base64

def chacha20_poly1305_encrypt(plaintext, key, associated_data=b""):
    # key should be 32 bytes (256-bit)
    cipher = ChaCha20_Poly1305.new(key=key)
    cipher.update(associated_data)
    ciphertext, tag = cipher.encrypt_and_digest(plaintext.encode())
    # Store nonce (12 bytes), tag (16 bytes) and ciphertext
    return base64.b64encode(cipher.nonce + tag + ciphertext).decode()

def chacha20_poly1305_decrypt(encrypted_text, key, associated_data=b""):
    data = base64.b64decode(encrypted_text)
    nonce = data[:12]      # ChaCha20_Poly1305 uses a 12-byte nonce
    tag = data[12:28]      # 16-byte tag
    ciphertext = data[28:]
    cipher = ChaCha20_Poly1305.new(key=key, nonce=nonce)
    cipher.update(associated_data)
    plaintext = cipher.decrypt_and_verify(ciphertext, tag)
    return plaintext.decode()

# Example usage
key = get_random_bytes(32)  # 256-bit key
plaintext = "HELLO123"
associated_data = b"AUTHENTICATED_MESSAGE"

encrypted = chacha20_poly1305_encrypt(plaintext, key, associated_data)
decrypted = chacha20_poly1305_decrypt(encrypted, key, associated_data)

print("ğŸ” Encrypted (ChaCha20-Poly1305):", encrypted)
print("ğŸ”“ Decrypted:", decrypted)
````

````python
from Crypto.Cipher import ChaCha20_Poly1305
import base64
import os

# Function to encrypt using XChaCha20-Poly1305
def xchacha20_poly1305_encrypt(plaintext, key, associated_data=b""):
    key = key[:32]  # Ensure the key is 256-bit (32 bytes)
    nonce = os.urandom(24)  # Generate a 192-bit (24-byte) random nonce
    cipher = ChaCha20_Poly1305.new(key=key, nonce=nonce)  # Initialize cipher with extended nonce
    cipher.update(associated_data)  # Add associated data (optional)
    ciphertext, tag = cipher.encrypt_and_digest(plaintext.encode())  # Encrypt & generate tag
    return base64.b64encode(nonce + ciphertext + tag).decode()  # Store nonce + ciphertext + tag

# Function to decrypt using XChaCha20-Poly1305
def xchacha20_poly1305_decrypt(encrypted_text, key, associated_data=b""):
    key = key[:32]  # Ensure the key is 256-bit (32 bytes)
    encrypted_data = base64.b64decode(encrypted_text)
    nonce = encrypted_data[:24]  # Extract 24-byte nonce
    ciphertext = encrypted_data[24:-16]  # Extract ciphertext
    tag = encrypted_data[-16:]  # Extract 16-byte authentication tag
    cipher = ChaCha20_Poly1305.new(key=key, nonce=nonce)  # Reinitialize cipher
    cipher.update(associated_data)  # Verify associated data (optional)
    decrypted_text = cipher.decrypt_and_verify(ciphertext, tag)  # Decrypt & authenticate
    return decrypted_text.decode()

# Example Usage
key = os.urandom(32)  # Generate a random 256-bit (32-byte) key
plaintext = "HELLO123"
associated_data = b"AUTHENTICATED_MESSAGE"  # Optional authenticated data

encrypted = xchacha20_poly1305_encrypt(plaintext, key, associated_data)
decrypted = xchacha20_poly1305_decrypt(encrypted, key, associated_data)

print("ğŸ” Encrypted (XChaCha20-Poly1305):", encrypted)
print("ğŸ”“ Decrypted:", decrypted)
````
